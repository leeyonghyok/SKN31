# LLM 개요

- **LLM**(**Large Language Model, 대형 언어 모델**)은 대량의 텍스트(최근에는 이미지·음성 등 멀티모달 데이터까지)에서 언어의 패턴을 학습해, 입력 문맥에 맞는 출력을 생성하는 AI 모델을 말한다.
  -  LLM은 언어를 다루는 범용 두뇌 역할을 하는 AI 모델이다.
  -  '사실 DB'가 아니라 '언어 패턴'을 잘 맞추는 두뇌이기 때문에, **검증/근거** 설계가 중요
- 대부분의 LLM은 **트랜스포머(Transformer)** 아키텍처에 기반해 설계 되었으며 Layer를 아주 많이 연결한 큰 구조를 가지고 있다. 이를 통해 방대한 양의 텍스트 데이터를 학습하여 인간의 언어를 이해하고 생성하는 능력을 가진다.
- 다양한 분야의 대량의 데이터로 학습했기 때문에 특정 분야에 국한되지 않고 다양한 작업에 바로 사용할 수있다.

## LLM 모델의 활용방식의 변화

### 초기(2022~2023)

- 사용자가 질문하면 LLM이 답하는 서비스.
- 그럴듯한 답변 생성이 주요 목표였다.
- 원하는 응답을 얻고 할루시네이션을 줄이기 위해 프롬프트 설계가 중요했다.
  - 프롬프트 엔지니어링이 중요했다.  
- 한 번의 호출로 끝나는 단발성 작업 중심.

### 2024년

- RAG(Retrieval-Augmented Generation)가 대중화되었다.
- 수학, 코딩과 같이 논리적 사고가 필요한 일에 특화된 추론(Reasoning) 모델이 등장하였다. (2024년 12월 OpenAI o1이 발표됨)
- Agent 프레임워크와 워크플로우 기반 개발 시작되었다.
- 단순 프롬프트 설계보다 시스템 설계의 중요성이 증가하기 시작했다.

### 2025년 이후
본격적인 AI Agent의 시대가 시작됨.
이 시기부터는 LLM을 단순 채팅 엔진이 아니라 작업을 수행하는 실행 주체로 활용하는 흐름이 강해졌다.

- 도구 사용(Tool Use / Function Calling)
  - 검색, DB 조회, 계산, API 호출 등의 외부 도구를 활용한다.

- Agent와 Workflow
  - 여러 단계를 반복하거나 분기하면서 목표를 수행한다.

- 장문 컨텍스트(Long Context)
  - 대부분의 모델들이 100만 토큰 이상의 입력(context window)을 처리할 수 있게 되었다. 
  - 문서, 코드베이스, 로그 등을 통합 분석하는 활용이 증가하였다.

- 모델 선택(Model Routing)
  - 속도, 비용, 추론 능력에 따라 여러 모델을 조합하여 사용한다.

- 관측성(Observability)과 평가(Evaluation)
  - 응답 품질뿐 아니라 처리 과정 전체를 추적·평가·모니터링하는 체계가 필수 요소가 되었다.

- 컨텍스트 엔지니어링(Context Engineering)
  - 프롬프트 자체보다 모델에 어떤 정보를 어떤 방식으로 제공할 것인가가 중요해졌다.


## LLM의 주요 특징

- **대규모 모델로 대규모 데이터 학습**
    - LLM은 두가지 측면에서 대규모(Large)이다.
        - 모델의 크기
            - LLM은 수십억에서 수조 개에 이르는 매개변수를 보유하고 있다.
        - 훈련 데이터셋
            - GPT-3 는 4990억 토큰의 데이터셋을 학습함. (보통 1토큰이 대략 영문 4글자 정도됨.)
    - 일반적으로 **70억(7B)개** 이상의 파라미터를 가진 모델이 LLM으로 간주되며, 대표적으로 **GPT-3.5**는 **1750억(175B)** 개의 파라미터를 갖고 있다.
    - 이러한 대규모 학습 덕분에 문맥을 이해하고 연관된 문장을 예측하는 성능이 높아진다.
- **자연어 생성**
    - LLM은 프롬프트에 기반해 자연스럽고 맥락을 고려한 언어를 생성할 수 있다.
    - 특정 분야에 국한되지 않고 다양한 작업에 적용 가능하여 범용성이 좋다. 
    - 이는 **챗봇, 자동 텍스트 생성, 요약, 번역** 등 다양한 응용 분야에서 활발히 활용되며, 창작과 데이터 분석의 도구로도 점점 널리 사용되고 있다.
    - 예) 질문에 대한 답변 생성, 스토리텔링, 정보 요약 등 다양한 형태의 텍스트 응답 생성이 가능하다.
- **다국어 지원**
    - LLM은 다양한 언어로 학습되며, 이를 통해 **언어 간 번역** 및 **다국어 텍스트 처리**에서도 우수한 성능을 보인다.
    - 이와 같이 여러 언어를 지원하는 모델을 **Multilingual Model**이라고 한다.
    - 다국어 지원을 통해 다양한 국가의 사용자에게 폭넓은 접근성을 제공하며, 언어 장벽을 줄이는 데 기여한다.
- **파운데이션 모델**
    - LLM 모델은 범용적인 특징을 가지며, 다양한 분야에 적용할 수 있도록 설계되어 있다. 이는 대규모 데이터를 활용해 학습된 모델이기 때문이다.
    - 이런 범용적인 특징으로 LLM을 파운데이션(기반) 모델이라고 한다.
  

  

## LLM의 주요 특징

* **대규모 모델과 대규모 데이터 학습**

  * Large Language Model(LLM)은 **대규모 파라미터를 가진 모델이 대규모 데이터로 사전학습되어 다양한 언어 작업에 범용적으로 사용될 수 있는 모델**을 말한다.
  * LLM의 **Large**는 크게 두 가지 의미를 가진다.

    * **모델의 크기**

      * LLM은 수십억에서 수천억 개 이상의 **파라미터**를 가진다.
      * 일부 최신 모델은 전체 파라미터 수가 1조 개를 넘는다고 공개되기도 한다.
  
    * **학습 데이터의 규모**

      * LLM은 웹 문서, 책, 논문, 코드, 설명문, 대화 데이터 등 매우 큰 데이터를 **토큰** 단위로 학습한다.
        * 토큰은 모델이 텍스트를 처리하는 기본 단위이다. 영어 기준으로는 보통 1토큰이 약 3~4글자 또는 약 0.75단어 정도에 해당한다. 한국어는 대략 1토큰이 1 ~ 2글자 정도 된다. (한글의 토큰 소비량이 더 크다.)
      * **이러한대규모 학습 덕분에 LLM은 단어와 문장 사이의 관계, 문맥, 표현 방식, 지식 패턴을 학습하고 다음에 올 가능성이 높은 토큰을 예측할 수 있다.**

  * 예를 들어 초기 대표 모델인 **GPT-3**는 약 **1,750억 개의 파라미터**를 가진 모델이었다. 이후 공개 모델들은 더 큰 학습량과 다양한 구조를 사용한다.

    * **Meta Llama 3.1**: 8B, 70B, 405B 모델이 공개되었고, 약 **15조 토큰** 규모로 학습되었다.
    * **DeepSeek-V3**: 전체 671B 파라미터를 가진 MoE 모델이며, 한 번에 전체 파라미터를 모두 쓰는 것이 아니라 토큰마다 약 37B 파라미터를 활성화한다. 학습 토큰 수는 약 **14.8조 토큰**으로 공개되었다.
    * **Qwen3 계열**: 0.6B부터 235B 규모의 모델이 있으며, 약 **36조 토큰** 규모로 학습되었다. Qwen3-Max는 전체 파라미터가 1조 개를 넘는 모델로 공개되었다.
    * **GPT, Claude, Gemini** 같은 상용 최신 모델들은 성능과 기능은 공개하지만, 파라미터 수나 전체 학습 토큰 수를 자세히 공개하지 않는 경우가 많다.

  > * 토큰 규모를 책으로 비유하면, 영문 300쪽 분량의 책 1권을 약 10만 단어, 즉 약 13만 토큰으로 볼 수 있다.
  >   * 15조 토큰은 대략 1억 권 이상, 36조 토큰은 대략 2억 7천만 권 정도의 책에 해당하는 규모로 볼 수 있다.


* **자연어 생성과 이해**

  * LLM은 사용자가 입력한 **프롬프트**를 바탕으로 자연스럽고 맥락에 맞는 문장을 생성할 수 있다.
  * 단순히 문장을 이어 붙이는 것이 아니라 질문의 의도, 앞뒤 문맥, 지시사항, 예시를 참고해 응답을 만든다.
  * 그래서 챗봇, 문서 작성, 요약, 번역, 코드 작성, 데이터 분석 보조, 아이디어 생성 등 다양한 자연어 처리 작업에 활용된다.

* **다국어 지원**

  * 최신 LLM은 영어뿐 아니라 한국어, 중국어, 일본어, 스페인어 등 다양한 언어를 처리할 수 있다.
  * 여러 언어를 함께 학습한 모델을 일반적으로 **다국어 모델**이라고 한다.
  * 다만 언어별 성능은 동일하지 않다. 학습 데이터가 많은 영어에서는 성능이 높은 편이고, 상대적으로 학습 데이터가 적은 언어에서는 성능 차이가 날 수 있다.
  * 2026년 기준의 주요 LLM들은 단순 번역을 넘어 다국어 질의응답, 문서 요약, 코드 설명, 검색 보조 등에 폭넓게 활용된다.

* **파운데이션 모델**

  * LLM은 특정 작업 하나만을 위해 만들어진 모델이 아니라, **다양한 범용적인 작업에 재사용할 수 있는 파운데이션 모델** 이다.
  * 하나의 LLM을 기반으로 챗봇, 검색 보조, 코딩 도우미, 고객 상담, 문서 분석, 교육용 튜터 같은 여러 응용 서비스가 만들어질 수 있다.
  * 최근에는 텍스트만 처리하는 모델을 넘어 이미지, 음성, 영상, 문서 파일까지 함께 처리하는 **멀티모달 파운데이션 모델**로 발전하고 있다.
  * 최신 LLM은 단순히 모델을 크게 만드는 것뿐 아니라 **학습 데이터 품질, 추론 능력, 긴 문맥 처리, 도구 사용, 멀티모달 처리, 비용 효율성**까지 함께 개선하는 방향으로 발전하고 있다.
  * 따라서 LLM은 “매우 큰 텍스트 생성 모델” 정도로만 이해하기보다, **대규모 데이터로 학습되어 자연어를 이해·생성하고 다양한 지식 작업의 기반으로 쓰이는 범용 인공지능 모델**이라고 할 수  있다.


## LLM의 한계

* **편향(Bias) 문제**
    * LLM은 웹 문서, 책, 코드, 뉴스, 커뮤니티 글 등 다양한 데이터를 학습한다.
    * LLM은 학습한 텍스트 데이터에 포함된 편향된 정보를 반영할 수 있어, **차별적 표현**이나 **정치적 편향**을 포함한 출력을 생성할 위험이 있다.
    * 이런 편향 문제를 줄이기 위해 학습 데이터 정제, 안전성 학습, 편향 평가, 레드팀 테스트, 사용자 피드백 반영 등이 사용되지만, 편향을 완전히 제거하기는 어렵다.
* **잘못된 정보 생성 (할루시네이션)**
    * LLM은 사용자의 질문에 대해 그럴듯한 문장을 생성하는 능력이 뛰어나지만, 항상 사실만 말하는 것은 아니다.
    * LLM은 기본적으로 언어 패턴과 문맥을 바탕으로 다음에 올 가능성이 높은 토큰을 예측하는 방식으로 동작한다.
      따라서 실제로 존재하지 않는 사실, 잘못된 수치, 가짜 논문, 존재하지 않는 출처, 틀린 코드 설명을 마치 확실한 것처럼 생성할 수 있다.
    * 이런 현상을 **할루시네이션(Hallucination)** 이라고 한다.
    * 교육, 의료, 법률, 금융, 채용, 공공 의사결정처럼 사람에게 큰 영향을 주는 영역에서는 사람이 최종 판단을 해야 한다.
* **높은 에너지 소비**
    * LLM은 학습과 실행 과정에서 많은 GPU, 데이터센터, 전력, 냉각 자원을 필요로 한다.
    * 과거에는 모델을 한 번 학습시키는 데 드는 비용이 주로 문제로 언급되었지만, 2026년 이후로 많은 사용자가 매일 모델을 호출하는 추론(Inference) 비용이 훨씬 더 커지고 있다.
* **프라이버시 및 보안 이슈**
    * LLM이 다양한 공공 데이터로 학습되는 만큼, 민감한 정보가 노출될 가능성이 존재한다.
    * **데이터 보호**와 **프라이버시 강화**를 위한 기술이 필요하며, 민감한 정보를 학습하지 않도록 하는 규제 방안도 연구되고 있다.

## Open Source LLM과 Closed Source LLM

### Open Source LLM
- 소스 코드와 모델 가중치가 공개되어 누구나 접근, 수정, 배포할 수 있는 대규모 언어 모델.
- 연구자, 개발자, 기업 등이 자유롭게 활용하고 커스터마이징 가능.
- **주요 모델**
  - **Llama 시리즈**: Meta에서 개발한 모델로, 2025년 4월 멀티모달을 지원하는 Llama 4 시리즈(Scout, Maverick, Behemoth)까지 발표되었다.
  - **DeepSeek 시리즈**: 중국 스타트업 DeepSeek에서 개발한 모델. 2025년 초 R1 모델이 주요 상용 LLM과 대등한 성능을 낮은 비용으로 달성해 주목받았으며, V3, R1, V4 등이 오픈소스로 공개되어 있다.
  - **Qwen 시리즈**: Alibaba에서 개발한 모델. Qwen3 시리즈까지 공개되었으며, 0.6B부터 235B 규모의 다양한 모델이 있다.
  - **Mistral 시리즈**: 프랑스 Mistral AI에서 개발한 모델. 2025년 12월 Mistral 3 시리즈가 발표되었으며, 대표 모델인 Mistral Large 3는 Apache 2.0 라이선스로 공개되어 있다.
  - **Gemma 시리즈**: Google DeepMind에서 제공하는 오픈소스 LLM. Gemma 3(2025년 3월), Gemma 4(2026년 4월)까지 발표되었다.
  - **Solar**: 국내 기업 업스테이지(Upstage)에서 개발한 LLM.

### Closed Source LLM
- 상용 LLM.
- 소스 코드와 모델 가중치가 비공개되어 특정 기업이나 기관만이 접근 및 활용할 수 있는 대규모 언어 모델.
- 일반 사용자나 개발자는 API 등을 통해 제한적으로 기능을 이용할 수 있음.
- **주요 모델**
  - **GPT 시리즈**: OpenAI에서 개발한 모델.
  - **Claude 시리즈**: Anthropic에서 개발한 모델.
  - **Gemini 시리즈**: Google DeepMind에서 개발한 모델.
  - **Grok 시리즈**: 일론 머스크가 설립한 xAI에서 개발한 모델. 현재 Grok 3까지 출시되었으며 API를 통해 이용 가능하다.


# [LangChain](https://www.langchain.com/)과 LangChain 생태계

![langchain_logo.png](figures/langchain_logo.png)

## LangChain이란?

LangChain은 대규모 언어 모델, 즉 LLM을 기반으로 애플리케이션과 에이전트를 쉽게 개발할 수 있도록 도와주는 오픈 소스 프레임워크이다.
LLM 자체는 사용자의 질문에 답변하거나 문장을 생성할 수 있지만, 실제 서비스에서는 단순히 모델을 한 번 호출하는 것만으로는 정확하고 충실한 답변을 받을 수 없는 경우가 많다.   
특히 AI Agent개발을 위해서는 더욱 많은 기능들이 LLM 호출 전 후로 실행되어야 한다.   
예를 들어 다음과 같은 기능들이 필요하다.

* 사용자의 질문을 프롬프트로 구조화하기
* OpenAI, Anthropic, Google Gemini, Ollama 등 다양한 모델을 연결하기
* 검색 시스템, 데이터베이스, API, 파일, 웹 검색 도구와 연결하기
* 이전 대화 내용을 기억하기
* 모델 출력 결과를 JSON 같은 구조화된 형식으로 받기
* 에이전트가 상황에 따라 도구를 선택하고 반복적으로 작업하게 만들기
* 실행 과정에서 오류가 발생했을 때 디버깅하고 평가하기

LangChain은 이러한 기능을 직접 처음부터 구현하지 않아도 되도록, LLM 애플리케이션 개발에 필요한 구성 요소를 제공한다.

LangChain은 2022년 Harrison Chase가 오픈 소스 프로젝트로 시작였고 이후 빠르게 생태계가 확장되었고, 2024년 1월 최초의 안정화 버전인 v0.1이 발표되었다.
2025년 10월에는 LangChain v1.0이 발표되면서 에이전트 개발을 위한 production-ready 기반으로 구조가 정리되었다.


## LangChain 특징

LLM 애플리케이션은 일반적인 웹 애플리케이션과 다른 특징을 가진다.

일반적인 프로그램은 같은 입력이 들어오면 대체로 같은 절차를 따라 같은 결과를 만든다. 반면 LLM 기반 애플리케이션은 모델의 응답이 확률적이고, 사용자의 입력도 매우 다양하다. 또한 모델이 스스로 외부 도구를 호출하거나, 검색 결과를 읽고 답변하거나, 여러 단계의 작업을 수행해야 하는 경우가 많다.    
예를 들어 "우리 회사 휴가 규정을 알려줘"라는 질문에 답하는 챗봇을 만든다고 하자. 이때 단순히 LLM에게 질문만 전달하면 정확한 답을 보장하기 어렵다.    
회사 규정 문서를 검색하고, 관련 내용을 찾아서, 그것을 기반으로 답변해야 한다.

이런 구조는 보통 다음과 같은 흐름을 가진다.

```text
사용자 질문
   ↓
질문 분석
   ↓
관련 문서 검색
   ↓
검색 결과를 프롬프트에 삽입
   ↓
LLM 호출
   ↓
답변 생성
   ↓
출력 형식 정리
```

LangChain은 이와 같은 흐름을 효과적으로 구성하기 위한 여러 기능을 제공한다. 특히 RAG, 챗봇, 문서 검색, 에이전트, 자동화 워크플로우를 만들 때 자주 사용된다.



## LangChain의 주요 구성 요소

## Model

* Model은 실제 LLM 호출을 담당한다. 
* LangChain은 여러 제공자(OpenAI GPT, Antrophic Claude, Google Gemini, Huggingface LLM 등)들의 모델을 공통된 방식으로 사용할 수 있게 해준다. 모델 제공자를 바꿔도 애플리케이션의 전체 구조를 크게 바꾸지 않고 사용할 수 있다.

## Prompt

* Prompt는 LLM에게 전달하는 지시문이다.
* 단순히 사용자의 질문만 전달하는 것이 아니라, 시스템의 역할, 답변 형식, 참고할 자료/문서, 제약 조건 등을 다양한 내용을 질문과 함께 전달한다.
* LangChain은 이러한 프롬프트를 템플릿으로 만들고, 실행 시점에 필요한 값을 채워 넣어 Prompt를 쉽게 만들 수있도록 한다.

## Chain
* LLM 애플리케이션에서는 사용자 질문을 받아 최종 답변을 생성하기까지 여러 작업이 단계적으로 수행된다.
   * 프롬프트 생성 → 모델 호출 → 출력 파싱
   * 질문 입력 → 문서 검색 → 검색 결과를 포함한 프롬프트 생성 → 모델 호출 → 답변 생성
* 이렇게 여러 작업을 연결한 실행 흐름을 일반적으로 workflow라고 한다.
* LangChain에서 **Chain은 이러한 workflow를 구성하는 방식**이다.
* Langchain은 이런 workflow를 `LCEL` (LangChain Expression Language) 이용하여 쉽게 구현할 수 있다.
* Chain은 하나의 작업만 수행하는 것이 아니라 여러 구성 요소를 연결하여 입력을 처리하고 최종 출력을 만드는 실행 파이프라인이라고 할 수있다.
* 복잡한 workflow나 agent 개발을 위하여 Langgraph Framework을 제공한다.
> LangGraph는 LangChain 생태계의 상태 기반 workflow 및 agent를 구성하는 프레임워크이다.

## Tool(도구)

* Tool은 LLM이 사용할 수 있는 외부 기능을 구현한 함수나 API이다.
* 다음과 같은 것들이 Tool이 될 수 있다.
  * 웹 검색
  * 데이터베이스 조회
  * 계산기
  * 프로그램 코드 실행
  * 구글 드라이브 조회
  * 파일 검색
  * 사내 API 호출
  * 벡터 데이터베이스 검색

* LLM을 뇌라고 한다면 Tool은 팔, 다리라고 할 수있다. 어떤 일을 할 지 LLM 이 판단하고 명령을 내리면 Tool이 실행한다.

## Retrieval

* Retrieval은 사용자의 질문과 관련 있는 정보를 외부 문서나 데이터베이스에서 찾아오는 기능이다.
* LLM은 학습 데이터에 없는 최신 정보(학습 이후에 발생한 정보)나 특정 회사에 공개되지 않은 사내 문서의 내용은 알 수 없다. 따라서 문서 검색 시스템과 결합해야 한다. 이때 자주 사용되는 구조가 RAG이다.
* RAG는 Retrieval-Augmented Generation의 약자로, "검색으로 보강된 생성"이라는 뜻이다.
* RAG의 기본 흐름은 다음과 같다.

```text
사용자 질문
   ↓
질문과 관련된 문서 검색
   ↓
검색된 문서를 LLM 프롬프트에 포함
   ↓
문서 기반 답변 생성
```

* LangChain은 문서 로더, 텍스트 분할기, 임베딩 모델, 벡터 스토어, 리트리버 등을 연결하여 RAG 시스템을 만들 수 있도록 지원한다.

## Memory

* Memory는 대화의 이전 내용을 기억하는 기능이다.
* 일반적인 LLM 호출은 독립적인 요청이기 때문에 이전 대화 내용을 따로 전달하지 않으면 모델은 이전 대화를 알 수 없다.
* LangChain과 LangGraph에서는 대화 이력, 사용자 정보, 작업 상태 등을 저장하고 다음 호출할 때 프롬프트에 넣어 전달하여 LLM이 지금까지의 대화 내용을 알 수 있게 한다.

* 메모리는 크게 다음과 같이 나눌 수 있다.

| 구분     | 설명                                 |
| ------ | ---------------------------------- |
| **상태 관리**  | 에이전트가 어떤 단계까지 작업했는지 저장             |
| **단기 메모리** | 현재 대화 세션 안에서 이전 메시지를 기억            |
| **장기 메모리** | 여러 세션에 걸쳐 사용자 선호, 과거 작업, 중요 정보를 저장 |

## Agent

* Agent는 LLM이 단순히 한 번 답변하는 것이 아니라, 목표를 달성하기 위해 여러 단계를 수행하는 구조이다. 특히 목표 달성여부에 대한 판단을 LLM이 하고 그결과에 따라 계속 다음 작업을 진행할 지 멈출지를 결정한다.
* 에이전트는 다음과 같은 반복 구조를 가진다.

```text
사용자 요청
   ↓
LLM이 다음 행동 결정
   ↓
필요하면 도구 호출 ←───────┐
   ↓                       │
도구 결과 확인             │
   ↓                       │
다시 LLM이 다음 행동 결정 ─┘
   ↓
최종 답변 생성
```

* LangChain은 `create_agent`와 같은 기능을 통해 모델, 도구, 시스템 프롬프트를 연결하여 에이전트를 만들 수 있게 한다.



# LangChain 생태계

![langchain ecosystem](figures/Langchain_ecosystem.png)

* LangChain 생태계는 크게 다음 세 가지 축으로 이해할 수 있다.
  * LangChain: LLM 애플리케이션과 에이전트 **개발 프레임워크**
  * LangGraph: 복잡한 에이전트 흐름을 제어하는 상태 기반 **에이전트 오케스트레이션 프레임워크**
    * **에이전트 오케스트레이션**란 에이전트 구성 요소들을 구성하고 조율하고 제어하는 작업을 수행하는 것을 말한다.
  * LangSmith: 실행 추적, 디버깅, 평가, 배포, 모니터링 플랫폼


각 도구의 역할은 다르지만 서로 보완적으로 사용된다.

## LangChain

* LangChain은 LLM 애플리케이션을 빠르게 만들기 위한 상위 수준의 프레임워크이다. LLM 어플리케이션을 만드는데 필요한 다양한 기능들을 제공하는 생태계의 가장 기본 바탕(Base)가 되는 프레임워크이다.
* 주요 역할
  * 모델 호출 추상화
  * 프롬프트 구성
  * 도구 연결
  * 구조화된 출력 처리
  * 검색 기반 생성, 즉 RAG 구성
  * 에이전트 생성
  * 다양한 외부 서비스와 통합

* LangChain은 특히 다음과 같은 경우에 적합하다.
  * 간단한 챗봇을 만들 때
  * RAG 기반 문서 질의응답 시스템을 만들 때
  * LLM에게 도구를 연결하고 싶을 때
  * 빠르게 프로토타입을 만들고 싶을 때
  * 모델 제공자를 바꿀 가능성이 있을 때
  * 에이전트 구조를 간단히 구현하고 싶을 때

## LangGraph

* LangGraph는 LangChain 생태계에서 더 복잡한 에이전트나 워크플로우를 만들기 위한 프레임워크이다.
* LangChain이 "빠르게 만들기 위한 프레임워크"라면, LangGraph는 "정교하게 제어하기 위한 프레임워크"라고 볼 수 있다.
* LangGraph는 이름처럼 그래프 구조를 사용하며 분기, 반복 등의 복잡한 흐름을 구성할 수 있다.

* 예를 들어 고객 지원 에이전트를 만든다고 할 때 이 에이전트는 다음과 같은 흐름을 가질 수 있다.

```text
문의 접수
   ↓
문의 유형 분류
   ↓
주문 관련 문의인가?
   ├─ 예: 주문 DB 조회
   └─ 아니오: 일반 FAQ 검색
   ↓
답변 초안 생성
   ↓
위험한 작업인가?
   ├─ 예: 담당자 승인 요청
   └─ 아니오: 사용자에게 답변
```

* 이런 흐름은 분기, 반복, 상태 저장, 승인 절차등이 필요하기 때문에 단순한 순차 체인으로 만들기 어렵다.
* LangGraph는 이런 복잡한 흐름을 그래프 형태로 명확하게 정의할 수 있게 해준다.

* LangGraph는 다음과 같은 경우에 적합하다.
  * 복잡한 에이전트를 만들 때
  * 여러 단계의 워크플로우가 필요할 때
  * 조건 분기와 반복이 많을 때
  * 실행 중간 상태를 저장해야 할 때
  * 사용자의 승인 후 다음 단계로 진행해야 할 때
  * 장시간 실행되는 작업을 안정적으로 처리해야 할 때
  * 프로덕션 수준의 에이전트가 필요할 때

## LangSmith
* LangSmith는 LLM 애플리케이션과 에이전트를 관찰하고 평가하고 개선하기 위한 플랫폼이다.
* 구현한 LLM 챗봇이 이상한 답변을 했을 때 다음과 같은 부분을 확인해야 한다.
   * 사용자의 원래 질문은 무엇이었는가?
   * 어떤 프롬프트가 모델에 전달되었는가?
   * 어떤 문서가 검색되었는가?
   * 어떤 도구를 호출했는가?
   * 도구 호출 결과는 무엇이었는가?
   * 모델은 어느 단계에서 잘못 판단했는가?
   * 비용과 응답 시간은 얼마나 걸렸는가?
   * 이전 버전보다 성능이 좋아졌는가?
* LangSmith는 이러한 정보를 추적하고 분석할 수 있게 해준다.
