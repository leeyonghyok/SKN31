# Agent 개요

## AI Agent란?

- **AI Agent는 사용자의 목표를 달성하기 위해 대형 언어 모델(LLM)을 추론과 의사결정의 중심으로 사용하여 작업 계획 수립, 실행, 관찰, 판단 과정을 반복하며 자율적으로 작업을 수행하는 인공지능 시스템이다.**.**
- 일반적인 LLM 사용 방식은 다음과 같이 한 번의 질문에 대해 답변을 생성하는 구조이다.
  
    ```text
    사용자 질문 → LLM → 답변
    ```
- AI Agent는 단순히 답변만 생성하지 않는다. **사용자의 목표를 이해한 뒤, 목표를 달성하기 위해 필요한 작업을 스스로 판단하고 실행**한다.

    ```text
    목표 이해 → 계획 수립 → 도구 실행 → 결과 관찰 → 다음 행동 결정
    ```
    - **목표 이해.**
        - Agent는 사용자의 요청을 단순한 문장으로만 처리하지 않고, 사용자가 실제로 달성하려는 목표가 무엇인지 파악한다.
    - **작업 계획.**
        - 목표를 달성하기 위해 어떤 순서로 작업해야 하는지 판단하고, 필요한 단계를 나눈다.
    - **외부 도구 사용**
        - Agent는 필요에 따라 검색 도구, API, 데이터베이스, 파일 시스템, 계산기, 코드 실행 환경 등 외부 도구를 호출할 수 있다.
    - **실행 결과를 관찰하고 다음 행동을 결정.**
        - 도구 실행 결과가 충분한지, 추가 작업이 필요한지 판단한 뒤 다음 단계를 결정한다.
    - **반복 실행**
        - 목표가 달성될 때까지 필요한 과정을 반복한다.
    - 모든 단계를 사람이 직접 지시하지 않아도, Agent는 주어진 목표를 기준으로 스스로 판단하며 작업을 이어갈 수 있다.


- 예를 들어 사용자가 “최근 AI 반도체 시장 동향을 조사해서 요약해줘”라고 요청했다고 하자.
  
    ```text
    1. 목표 파악: AI 반도체 시장 동향 조사
    2. 계획 수립: 주요 기업, 시장 규모, 최근 이슈, 전망을 조사
    3. 도구 사용: 웹 검색, 데이터베이스 조회
    4. 결과 관찰: 수집한 자료가 충분한지 판단
    5. 추가 실행: 부족한 정보가 있으면 추가 검색
    6. 최종 답변: 조사 결과를 정리해 사용자에게 제공
    ```

이처럼 Agent는 단순 응답 생성이 아니라 **목표 달성을 위한 일련의 작업 과정**을 수행한다.

## Workflow와 Agent

- AI 시스템을 구현할 때 자주 비교되는 방식이 **Workflow**와 **Agent**이다.
-   **Workflow**
    - 미리 정해진 절차에 따라 실행되는 방식이다.
    - 실행 순서와 작업 단계가 사전에 고정되어 있으며, 입력이 들어오면 정해진 흐름대로 처리된다.
    - 반복적이고 정형화된 작업에 적합하다.
- **Agent**
    - 주어진 목표를 달성하기 위해 스스로 판단하면서 행동을 결정하는 방식이다.
    - 처음부터 모든 실행 단계가 고정되어 있지 않고, 상황에 따라 필요한 작업과 도구를 선택하여 실행한다.
    - Agent는 작업 과정이 고정되어 있지 않거나, 중간 결과에 따라 다음 행동이 달라지는 문제에 적합하다.


## Agent의 주요 특징

1. **자율성** (Autonomy)
   - Agent는 **사전 정의된 규칙**(Rule-based)에 의존하지 않고, LLM을 통해 현재 상황을 인지하고 추론하여 스스로 결정을 내리고 다음 행동을 계획할 수 있다.
   - 즉, 사용자가 일일이 수행 단계를 지시하지 않아도, Agent는 **현재 상황을 해석하고 다음 행동을 동적으로 결정한다.**

2. **목표 지향성** (Goal-Oriented)
    - Agent는 단순히 질문에 답변하는 **단순한 대화 상대가 아니라, 특정 목표와 복잡한 작업을 달성하기 위해 설계된 시스템**으로 최종 목표에 도달할 때 까지 반복적으로 행동한다.
    - Agent는 어떤 처리를 할 때 항상 **"이 행동이 목표 달성에 도움이 되는가"를 기준**으로 판단한다.

3. **도구 활용** (Tool Utilization)
    - Agent는 대형 언어 모델의 언어 이해 능력, 사전 학습 정보에만 의존하지 않고, 다양한 **외부 도구와 API를 함께 활용하여 실제 작업을 수행**한다.
    - 이를 통해 LLM이 처리할 수없는 **최신 정보 검색, 복잡한 계산, 외부 시스템 제어**등과 같은 작업이 가능하다. 이를 통해 Agent는 **"말만 하는 AI"가 아니라 "실제로 일을 처리하는 AI"로 기능한다.**

4. **인간 개입 최소화** (Human-in-the-loop 최소화)
    - Agent는 **문제 분석, 의사 결정, 행동 실행의 대부분을 스스로 결정하고 수행하도록 설계된다.**
    - 사람은 **초기 목표 제시, 결과 확인 및 실행 최종 승인** 정도에 만 개입한다.

# 싱글에이전트(Single-Agent)와 멀티에이전트(Multi-Agent)

## 싱글에이전트(Single-Agent)
- **하나의 에이전트가 작업의 처음부터 끝까지 전 과정을 책임지는 구조**로 구성된 Agent.
- 싱글에이전트는 **"인식 → 추론 → 행동(도구 호출) → 결과 관찰"** 흐름의 모든 판단이 **하나의 추론 컨텍스트(context) 안**에서 이루어진다. 즉, 도구(검색, 코드 실행, 파일 읽기 등)와 메모리를 모두 한 에이전트가 쥐고 있으며, 작업 전체의 맥락이 단일한 의사결정자 안에 통합되어 있다.
    - 예를 들어 "이 코드의 버그를 찾아 고쳐줘"라는 요청에서, 하나의 에이전트가 코드를 읽고, 원인을 추론하고, 수정안을 실행하고, 테스트 결과를 확인하는 일을 혼자 모두 처리한다.
- **장점:** 구조가 단순해 응답이 빠르고(저지연), 토큰 소비가 적어 비용이 낮으며, 동작이 한 곳에서 일어나기 때문에 디버깅과 추적이 쉽다.
- **한계:** 단일한 컨텍스트 윈도우에 작업 전체가 담겨야 하므로 정보량이 많아지면 한계에 부딪히고, 단계들을 순차적으로 처리하기 때문에 여러 갈래를 동시에 진행하는 **병렬 처리가 불가능**하다. 작업 규모가 커질수록 이 두 가지(컨텍스트 한계·순차 병목)가 성능의 제약이 된다.

## 멀티에이전트 (Multi-Agent)

- **역할이 나뉜 여러 에이전트들이 협력하여 하나의 작업을 처리하는 구조**이다.
- 보통 **오케스트레이터(orchestrator) 혹은 슈퍼바이저(supervisor)** 라 불리는 상위 에이전트가 큰 작업을 작은 하위 작업으로 분해하고, 각 분야에 특화된 **워커(worker) 에이전트**들에게 위임한다. 각 워커는 자신만의 독립된 컨텍스트와 도구, 역할(role)을 가지며, 작업을 마치면 결과를 상위 에이전트(슈퍼바이저)에게 보고한다. 상위 에이전트는 이 결과들을 종합해 최종 산출물(응답)을 만든다.
- 멀티에이전트 구성 패턴으로는 상위가 하위를 지휘하는 **슈퍼바이저/오케스트레이터-워커 패턴**, 단계별로 결과를 넘겨주는 **순차 파이프라인 패턴**, 에이전트들이 서로 동등하게 의견을 주고받는 **수평적 협업 패턴** 등이 있다.

### 멀티에이전트가 효과적인 이유

- Anthropic이 공개한 멀티에이전트 리서치 시스템 자료에 따르면, 리드 에이전트(Opus)와 서브에이전트(Sonnet)로 구성한 멀티에이전트가 단일 에이전트(Opus) 대비 내부 평가에서 **약 90% 이상 높은 성능**을 보였다.
- 이 성능 향상은 주로 **토큰 사용량**, 그리고 추론을 여러 개의 독립된 컨텍스트 윈도우로 분산시킨 능력과 강하게 연결되어 있었다. 실제로 성능 변동의 약 80%가 토큰 사용량에 영향을 받았고, 그 외에는 도구 호출 수와 모델 선택 등이 영향을 주었다.
- 토큰 사용량이란 하나의 작업을 끝낼 때 까지 모델이 읽고 쓴 토큰의 총합이다. 여기에는 입력받은 토큰들(검색결과, 도구 반환값, 읽어들인 문서, 이전 대화맥락등등)과 출력토큰들(모델이 생성한 추론, 도구 호출명령, 최종 답변등등)들이 있다.
- 토큰 사용량이 만다는 것은 그 만큼 **모델이 많은 정보를 들여다 보고, 많이 생락을 했다**는 것을 뜻한다.
- 즉 **멀티에이전트가 강한 이유는 결국 문제를 해결하는데 "더 많은 사고량(토큰)"을 병렬로 효율적으로 투입할 수 있기 때문**이다.

### 멀티에이전트의 비용과 위험

- **조율 비용(coordination overhead):** 에이전트 간 작업을 나누고 결과를 주고받는 데 추가 비용이 든다.
- **토큰·비용 급증:** 여러 에이전트가 각자 추론하므로 토큰 사용량이 크게 늘어난다. Anthropic 자료 기준으로 **일반 채팅 대비** 단일 에이전트는 약 4배, 멀티에이전트 리서치 시스템은 약 15배(십수 배)까지 토큰을 소비한다. (싱글에이전트를 기준으로  비교해 보면 멀티에이전트는 대략 4배 수준이다.)
- **오류 전파(error propagation):** 한 에이전트의 오류가 다음 단계로 쉽게 번진다.
- **디버깅 난이도:** 여러 주체의 비결정적(non-deterministic) 상호작용을 추적해야 하므로 재현과 디버깅이 까다롭다.
- **맥락 단절 위험:** 서브에이전트들이 맥락을 충분히 공유하지 못하면 서로 모순되거나 일관성 없는 결과를 산출하고, 이를 합치는 과정에서 오히려 품질이 떨어질 수 있다. 즉 멀티에이전트는 만능이 아니며, 작업들이 서로 긴밀하게 의존해 **맥락 공유가 필수적인 경우**라면 무리하게 쪼개기보다 싱글에이전트로 두는 편이 오히려 안정적이다.

## 어떤 에이전트를 언제 사용해야 하는가

- 가장 중요한 원칙은 **"싱글에이전트로 단순하게 시작하고, 한계에 부딪히면 멀티에이전트로 확장한다"** 이다.
- 멀티에이전트는  **비용과 복잡도가 크기 때문**에 처음부터 **멀티에이전트를 선택하는 것은 과잉설계**(**over-engineering**)가 될 수 있다.

### 싱글/멀티 에이전트 트레이드오프

| 기준 | 싱글에이전트 우위 | 멀티에이전트 우위 |
| --- | --- | --- |
| 지연시간(응답 속도) | ✅ 빠름 | 느림 |
| 비용(토큰) | ✅ 낮음 | 높음(채팅 대비 ~15배) |
| 디버깅·추적 | ✅ 쉬움 | 어려움 |
| 병렬 처리 | 불가 | ✅ 가능 |
| 전문 영역 분담 | 어려움 | ✅ 가능 |
| 대용량 컨텍스트 처리 | 한계 있음 | ✅ 분산 처리 가능 |
| 긴밀한 맥락 의존 작업 | ✅ 안정적 | 부적합 |

### 싱글에이전트가 적합한 경우

- 대부분의 실무 작업은 사실 싱글에이전트로 충분하다.
- 작업의 전체 맥락이 하나의 컨텍스트 윈도우 안에 들어가고, 단계들이 순차적으로 이어져 이전 단계의 결과가 다음 판단에 곧바로 영향을 줄 때 싱글에이전트가 유리하다.

### 멀티에이전트가 적합한 경우

- 작업을 여러 갈래로 **병렬 처리**할 수 있을 때.
- 서로 다른 **전문 영역**이 필요할 때(예: 검색 전담, 코드 작성 전담, 검증 전담).
- 정보량이 너무 많아 하나의 컨텍스트 윈도우로는 감당이 안 될 때.

# ReAct 패턴
- Agent 구현의 바탕이 되는 이론.

## 개요

ReAct는 **Reasoning**(추론)과 **Acting**(행동)을 결합한 AI Agent의 프롬프팅 패턴이다. 대형 언어 모델(LLM)이 단순히 답변을 생성하는 것이 아니라, 문제를 분석하고(Reasoning)하고 그에 따라 필요한 도구를 선택하여 실행하며(Acting), 결과를 관찰하고(Observation) 다음 추론에 반영하는 과정을 반복하면서 문제를 해결하는  Agent 설계 방식이다.

### ReAct의 핵심 개념

전통적인 프롬프팅 방식은 질문에 대해 즉시 답변을 생성한다. 하지만 **ReAct 패턴**은 다음과 같은 순환 구조를 따른다:

```
질문 → 생각(Thought) → 행동(Action) → (결과)관찰(Observation) → 생각 → 행동 →  관찰 → ... → 답변
```

- 예)
  - 질문: "현재 서울의 날씨와 미세먼지 농도를 알려줘":

    - **Thought(생각)**: 날씨 정보를 얻기 위해 날씨 API를 호출해야겠다
    - **Action(행동)**: WeatherAPI.get_weather("서울")
    - **Observation(관찰)**: 기온 15도, 맑음
    - **Thought(생각)**: 이제 미세먼지 정보도 필요하다
    - **Action(행동)**: AirQualityAPI.get_pm("서울")
    - **Observation(관찰)**: PM10 농도 20㎍/㎥, 좋음
    - **Answer(답변)**: 서울의 현재 날씨는 기온 15도로 맑고, 미세먼지 농도는 30㎍/㎥로 좋은 상태이다

이처럼 ReAct는 Agent가 **사고 과정을 명시적으로 드러내면서** 외부 도구를 활용하여 복잡한 문제를 해결할 수 있도록 한다.

### ReAct의 장점

1. **투명성**: Agent의 사고 과정을 추적할 수 있어 디버깅이 용이하다
2. **정확성**: 외부 지식과 도구를 활용하여 환각(hallucination)을 줄인다
3. **유연성**: 다양한 도구를 조합하여 복잡한 작업을 수행할 수 있다
4. **해석 가능성**: 각 단계의 추론 과정을 이해할 수 있다




## 동작 단계
ReAct 패턴은 크게 **5단계**의 순환 구조로 동작한다.

1. **질문 입력** (Question)
     - 사용자로부터 해결해야 할 질문이나 작업을 입력받는다.
     - 예시
    ```bash
     User Input: "2024년 노벨 물리학상 수상자는 누구이고, 그들의 주요 업적은 무엇인가?"
    ```

2. **사고 단계** (Thought)
     - 언어 모델이 사용자의 질문(요청)을 분석하고, 문제 해결을 위해 다음에 무엇을 해야 할지 추론한다.(계획을 세운다.) 
     - 이 단계에서 Agent는:
         - 문제를 이해하고 분해한다
         - 필요한 정보가 무엇인지 파악한다
         - 어떤 도구를 사용해야 할지 결정한다
         - 충분한 정보가 모였는지 판단한다

       ```bash
       Thought: 2024년 노벨 물리학상 정보는 내 학습 데이터에 없을 수 있다. 최신 정보를 검색해야 한다.
       ```

3. **행동 단계** (Action)
     - 사고 단계에서 결정된 계획을 바탕으로 실제 행동을 실행한다. 행동은 일반적으로 외부 도구(Tool) 호출한다.   

        ```bash
        Action: 검색엔진.검색("2024 노벨 물리학상 수상자")
        ```

4. **관찰 단계** (Observation)
     - 행동(Acttion)의 결과를 관찰하고 얻은 정보를 확인한다.

       ```bash
       Observation: 2024년 노벨 물리학상은 John Hopfield와 Geoffrey Hinton에게 수여되었다. 
       ```

 5. **반복 또는 종료**
       - 관찰한 정보가 사용자 질문에 대한 답변을 하는데 충분하지 않다면 2번의 사고 단계로 돌아가 추가 행동을 계획한다.(사고→행동→관찰 반복) 
       - 충분한 정보가 모였다면 그 정보를 바탕으로 최종 답변을 생성한다.

          ```bash
          Thought: 수상자 이름은 알았지만 구체적인 업적 내용이 필요하다.

          Action: 정보검색.Search("John Hopfield Geoffrey Hinton 노벨상 업적")

          Observation: Hopfield는 연상 메모리를 위한 Hopfield 네트워크를 개발했고, Hinton은 딥러닝의 역전파 알고리즘과 볼츠만 머신을 연구했다.

          Thought: 이제 질문에 답할 충분한 정보가 모였다. 답변을 생성한다.

          Action: Finish("2024년 노벨 물리학상은 John Hopfield와 Geoffrey Hinton에게 수여되었다. Hopfield는 연상 메모리를 위한 Hopfield 네트워크를 개발했고, Hinton은 딥러닝의 역전파 알고리즘과 볼츠만 머신 연구로 인공신경망 발전에 기여했다.")
          ```

# Tool과 Tool Calling

## Tool

### Tool이란

- **Tool은 하나의 기능을 처리하는 함수이다.**
    - Tool은 하나의 명확한 기능을 수행하도록 정의된 함수(Function) 형태의 외부 실행 수단이다.
    - 대형 언어 모델(LLM)은 본래 텍스트를 생성하는 모델이므로, 실제 계산, 검색, 파일 처리, 데이터베이스 조회와 같은 현실 세계의 작업을 직접 수행할 수 없다. 이런 실제 작업을 대신 수행하는 단위가 Tool이다.
- AI Agent는 내부적으로 **LLM을 추론 엔진으로 사용해 판단하고 Tool을 호출해서 실행**한다.

### Tool이 필요한 이유 — LLM의 두 가지 본질적 한계

LLM은 강력한 추론(reasoning) 능력을 갖추고 있지만, 두 가지 본질적인 한계를 가진다. Tool은 이 한계를 보완한다.

#### 한계 1 — **외부 세계를 읽지 못한다** (Read 제약)

LLM은 학습 시점에 고정된 내부 지식만 갖고 있으며, 실행 시점의 외부 정보에 스스로 접근할 수 없다.

- **Knowledge Cutoff:** 학습 데이터가 마지막으로 업데이트된 시점 이후에 발생한 사건이나 정보를 알지 못한다.
- **실시간 데이터 미접근:** 현재 주가, 날씨, 데이터베이스 레코드처럼 매 순간 바뀌는 데이터는 학습 데이터 자체가 존재하지 않는다.
- **개인·조직 데이터 미접근:** 사용자의 파일, 이메일, 사내 문서 등 공개되지 않은 데이터는 LLM이 본래 알 수 없다.

Tool은 Web Search, DB 조회, File Reader 등을 통해 이 Read 제약을 해소한다.

#### 한계 2 — **외부 세계에 영향을 줄 수 없다** (Write/Execute 제약)

LLM은 텍스트를 생성할 뿐, 그 결과를 실제 세계에 반영하거나 실행할 수 없다.

- **연산 부정확:** 복잡한 수치 계산은 LLM의 언어 생성 방식으로는 신뢰하기 어렵다.
- **코드 실행 불가:** 코드를 작성할 수는 있지만, 직접 실행해 결과를 얻지는 못한다.
- **시스템 연동 불가:** 이메일 전송, 예약, 결제, 코드 배포처럼 실제 세계에 변화를 일으키는 액션(action)을 수행할 수 없다.

Tool은 Code Executor, API 호출, 메시지 전송 함수 등을 통해 이 Execute 제약을 해소한다.

#### **LLM의 한계와 Tool의 대응 관계**

| LLM의 한계 | 분류 | Tool의 대응 | 예시 |
|---|---|---|---|
| Knowledge Cutoff | Read | 최신 정보 조회 | Web Search, News API |
| 실시간 데이터 미접근 | Read | 현재 데이터 조회 | DB Query, Weather API |
| 개인·조직 데이터 미접근 | Read | 파일·문서 접근 | File Reader, Email Reader |
| 연산 부정확 | Execute | 정확한 계산 위임 | Calculator, Spreadsheet |
| 코드 실행 불가 | Execute | 코드 실행 | Python REPL, Shell |
| 시스템 연동 불가 | Execute | 외부 시스템 액션 | REST API, 메시지 전송 |
| 상태(state) 비저장 | Read/Write | 영속적 저장소 활용 | Memory Tool, Vector DB |

## Tool Calling 개요
**도구 호출**(Tool Calling)은 LLM이 자체적으로 수행할 수 없는 작업을 외부 도구나 API를 활용하여 처리할 수 있게 하는 핵심 기능이다.  
이를 통해 **LLM은 자체 지식의 한계를 넘어서는 최신 정보, 정확한 계산, 외부 시스템 제어 등의 능력을 확보**할 수 있다.

- **도구 호출(Tool Calling)의 3단계**
    1. 도구 바인딩 (Tool Binding): **LLM에 사용가능한 tool들을 등록한다.** 
        - LLM에게 사용 가능한 도구 목록과 각 도구의 기능(설명) 및 필요한 입력 매개변수를 미리 정의하여 제공한다.
            - Tool 함수
            - Tool 이름
            - Tool이 수행하는 기능 설명
            - 입력 파라미터 type과 설명
    2. **사용자 질의가 들어오면 LLM이 도구 사용 여부를 판단**한다.
        - LLM 자체 추론으로 해결가능한 문제인지 Tool을 호출해야 처리할 수있는 문제 인지를 확인한다.
        - **Tool 사용이 필요한 문제** 라고 판단하면 
          - 도구를 호출 하는 것이 아니라 어떻게 호출해야 하는지 요청정보를 생성한다.
            - 어떤 Tool을 사용해야 하는지, 그 Tool에 어떤 값을 전달해야하는지를 응답으로 생성한다.
    3.  **시스템이 Tool 호출 요청정보에 따라 Tool을 실행하고 그 결과를 LLM 에게 전달**한다.

     ![toolcalling_concept](figures/toolcalling_concept.png)

- **TOOL Calling 예**
    - LLM이 수학 계산을 수행해야 할 때, 직접 계산하는 대신 미리 정의된 '계산' 도구를 호출하여 정확한 결과를 얻을 수 있다.
  	- LLM에 최신 정보를 요청하는 질문이 들어왔을 때 '검색' 도구나 'Database 연동' 도구를 사용해 최신 뉴스를 검색하거나, 특정 데이터베이스에서 정보를 조회하는 등의 작업을 수행할 수 있다.
  	- LLM 이 작성한 결과를 google docs와 같은 문서도구나 SNS에 등록한다.

## LLM ToolCalling 흐름

### 1. Tool 생성 및 LLM Model에 tool binding
- `LLM-Model.bind_tools(tools=[tool_1, tool_2, ...])`

````python
from langchain_openai import ChatOpenAI
from langchain_exa import ExaSearchResults
# tool 과 llm 모델 생성
exa_search = ExaSearchResults(max_results=5)
model = ChatOpenAI(model="gpt-5.4-mini")

# llm 모델에 tool을 binding (tool binding)
tool_model = model.bind_tools(tools=[exa_search])
# model: LLM(tool X), tool_model: LLM + Tool들
````

````python
# TOOL 호출이 필요없는 질문에 대한 응답
result = tool_model.invoke("안녕하세요")
````

````python
# type(result)
print(result.content)   # LLM 응답 text
print(result.tool_calls)# 호출할 tool에 대한 정보(어떤 tool을 어떻게 호출할지)
# 응답에는 위 둘중 하나만 값이 있다.
````

````python
# Tool 호출이 필요한 질문
result2 = tool_model.invoke("2026년 7월 7일 서울의 날씨는 어떤가요?")
````

````python
print(result2.content)
print(result2.tool_calls)
````

````python
query = "2026년 7월 7일 서울, 런던, 뉴욕의 날씨를 알려줘. 그리고 오늘 월드컵 경기 결과들도 알려줘."
result3 = tool_model.invoke(query)
````

````python
result3.content
print(len(result3.tool_calls)) # list[dict]
for r in result3.tool_calls:
    print(r)
    print("------------------------------")
````

### 3. Tool의 처리(응답) 결과를 LLM 요청시 사용
- ToolMessage를 prompt 에 추가하여 LLM에 요청한다.
- ToolMessage 는 Tool Calling 정보를 가진 AIMessage 다음에 들어와야 한다.
- Prompt 순서
    1. 일반 prompt (system, 대화 history, .., human)
    2. AIMessage: tool calling 정보를 가진 AIMessage. (tool_model에 질의 받은 tool calling 정보가 있는 응답)
    3. ToolMessage:  Tool의 처리 결과

````python
from langchain_openai import ChatOpenAI 
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import chain
from langchain_tavily import TavilyResearch
from langchain_exa import ExaSearchResults

from datetime import date
from dotenv import load_dotenv 

load_dotenv(override=True)
````

````python
prompt_str = """
- 당신은 사용자의 다양한 질문에 답변하는 AI assistant입니다.
- 사용자의 질문에 상대적 시간 표현이 포함되어 있고 절대 날짜가 명시되어 있지 않은 경우, 오늘 날짜를 기준으로 실제 날짜를 계산한 뒤 검색합니다.
    - 상대적 시간 표현 예: 현재, 오늘, 어제, 내일, 이번 주, 지난주, 이번 달, 지난달, 올해, 작년
- 오늘 날짜는 {today}입니다.
- 답변에 필요한 정보가 부족하거나 최신 정보 확인이 필요한 경우, 적절한 tool을 사용해 정보를 수집합니다.
    - 질문에 답변하기 위해 웹 검색 tool을 사용할 때, 질문에 시점 정보가 필요한 경우 계산된 날짜나 시간 범위를 검색 쿼리에 포함해야 합니다.
- 웹 검색이 필요할 경우 binding 된 웹검색 tool(tavily_search, exa_search_results_json) 을 모두 사용합니다.
- 사용할 수 있는 tool은 다음과 같습니다.
    - tavily_search: 웹 검색 tool
    - exa_search_results_json: 웹 검색 tool"""
````

````python
# Chain구성
# 모델 정의
model = ChatOpenAI(model="gpt-5.4-mini")
# Tool 정의
tavily_search = TavilyResearch(max_results=5)
exa_search = ExaSearchResults(max_result=5)

# model-bind-tools
tool_model = model.bind_tools([tavily_search, exa_search])

today = date.today().strftime('%Y년 %m월 %d일')   # date -> str (2026년 07월 07일)
prompt = ChatPromptTemplate(
    [
        ('system', prompt_str), # SystemMessage
        ('human', "{query}"),   # HumanMessage
        MessagesPlaceholder(variable_name="tool_messages", optional=True)
        # Tool 호출결과인 ToolMessage 들을 넣을 placehold
        # AIMessage(tool 호출하라는 AI의 응답메세지) -> ToolMessage
    ],
    partial_variables={'today':today}

)
# prompt.invoke({"query":"질문"})

tool_model_chain = prompt | tool_model 

# tool_model_chain
## 질문 -[llm 판단:직접답변가능]->응답결과 (AIMessage.content)
## 질문 -[llm 판단:tool을 호출] ->Tool calls (AIMessage.tool_calls)
````

````python
tavily_search_name = tavily_search.name
exa_search_name = exa_search.name

@chain
def web_agent(query:str)->tuple[str, list]:
    """
    Agent의 전체 흐름을 처리하는 Runnable .
    사용자 질문을(query) 받아서  tool_model과 tool을 이용해 최종 응답을 생성해서 반한다.
    Args:
        query(str): 사용자 질의

    Returns:
        tuple[str, list]: AI의 최종답변(str)과 tool 메세지들(list)
    """
    # 0. ToolMessage들을 저장할 list
    messages = []
    # 1. tool model에 query를 질의
    ai_message = tool_model_chain.invoke(query) # ai_message.content="" 툴요청
    while ai_message.tool_calls: # 빈 리스트가 아니라면-> tool을 호출 [{}, {},{}]
        for tool_call in ai_message.tool_calls:
            if tool_call["name"] == tavily_search_name: 
                messages.append(tavily_search.invoke(tool_call))
            elif tool_call["name"] == exa_search_name:
                messages.append(exa_search.invoke(tool_call))

        # tool 호출완료 -> tool_model_chain(LLM)을 다시호출=> query, ToolMessage들
        messages = [ai_message] + messages 
        # AIMessage(tool calls정보를가진), ToolMessage(tool처리결과)

        ai_message = tool_model_chain.invoke(
            {"query":query, "tool_messages":messages}
        )
    # AI 최종 응답을 한 경우.(end of while)
    messages.append(ai_message) # 마지막 AI 응답을 포함
    return ai_message.content, messages
````

## @tool 사용
- 함수로 구현하고 `@tool` 데코레이터를 사용해 tool(StructuredTool)로 정의한다.
    - `langchain_core.tools` 모듈에 있다.
- tool name
    - 함수의 이름이 tool의 이름이 된다.
- parameters
    - 함수의 파라미터가 tool의 파라미터가 된다.
    - **type hint**를 이용해 타입을 지정한다.  
- description
    - doctring이 description이 된다.
    - RunnableBinding(model + tool)이 tool을 잘 찾을 수 있도록 하려면 **tool의 기능을 최대한 구체적**으로 작성한다.
- **@tool이 적용된 함수(StructuredTool)가 tool**이므로 model에 binding 한다.

````python
from langchain_core.tools import tool
# 사칙연산 툴을 정의
@tool
def plus(num1:int|float, num2:int|float) -> int|float:
    """받은 두개의 수를 덧셈하는 툴"""
    return num1 + num2

@tool
def minus(num1:int|float, num2:int|float) -> int|float:
    """받은 두개의 수를 뺄셈을 하는 툴"""
    return num1 - num2

@tool
def multiply(num1:int|float, num2:int|float) -> int|float:
    """받은 두개의 수를 곱셈을 하는 툴"""
    return num1 * num2

# 파라미터에 대한 자세한 설명을 해야 하는 경우. -> Pydantic.BaseModel로 정의
from pydantic import BaseModel, Field
class DivideToolParameter(BaseModel):
    num1: int|float = Field(..., description="나눌때 사용할 첫번째 값")
    num2: int|float = Field(..., description="나눌때 사용할 두번째 값. 0은 사용할 수 없다.")

@tool("divide_tool", args_schema=DivideToolParameter)
def divide(num1:int|float, num2:int|float) -> int|float:
    """받은 두개의 수를 나눗셈을 하는 툴"""
    return num1 / num2
# type(plus)
````

````python
# 위키백과사전을 검색해서 내용을 크롤링
from langchain_community.document_loaders import WikipediaLoader
from pydantic import BaseModel, Field
from langchain_core.runnables import chain
import json 

# Runnable 구현 -> tool 변환
@chain
def wikipedia_search(inputs: dict) -> str:
    """
    사용자의 쿼리를 검색어로 위키백과사전(wikipedia)를 검색하고 그 결과 중 k개의 문서내용을 반환한다. 
    Args:
        inputs(dict): key- query, max_result. query(str) 검색할 키워드, max_result(int) 검색할 결과수
    
    Returns:
        str - 검색한 결과들을 json 형식의 문자열로 반환.
              형식: {"result":[{검색된 문서정보들}, {검색된 문서정보들}]}
    """
    print("-------", inputs)
    query = inputs['query']
    max_result = inputs.get("max_result", 3)

    loader = WikipediaLoader(
        query=query, #검색어
        load_max_docs=max_result, # 검색할 문서 최대 개수
        lang="ko" # 검색할 wikipedia 나라.
    )
    cnt = 0
    result_list = []
    while True:
        cnt += 1
        print(f"{cnt}번째 시도")
        try:
            docs = loader.load() # list[Document]
            for doc in docs:
                result_list.append(
                    {
                        "content":doc.page_content, # 조회내용
                        "title": doc.metadata.get("title"),
                        "url": doc.metadata.get("source")
                    }
                )
            break # 정상적으로 종료 되면 break(while문 종료)
        except Exception as e: 
            # loader.load()에서 검색한 결과를 parsing할 때 문제가 발생. 발생여부가 불확실
            if cnt == 4: # 4번째 시도에서도 Exception 발생
                break # load 종료
            continue # 4번째가 아니면 다시 시도

    if result_list:
        result = {"result":result_list}
    else:
        result = {"result":"검색된 문서가 없습니다."}

    # List/Dictionary => str(JSON형식)
    return json.dumps(result, ensure_ascii=False)
````

## Vector Store(Vector 저장소)를 tool로 생성

    - **주요 파라미터**
        - `retriever`: vectorstore.as_retriever() 등으로 만든 Retriever
        - `name`: Agent가 사용할 Tool 이름
        - `description`: Agent가 언제 이 Tool을 써야 하는지 판단하는 설명

        - `document_separator`: 검색결과가 여러개일 경우 그 결과를 하나의 문자열로 묶어서 반환한다. 합칠때 사용할 구분자를 지정한다. 기본값은 `\n\n` 이다.

2. `@tool` 을 이용해 직접 구현

3. Retriever도 Runnable이므로 `as_tool()` 메소드를 이용해 변환할 수 있다.

### text loading -> Document 생성
- 레스토랑 메뉴를 vector store에 저장한다.
1. 메뉴 text 를 로딩한다.
2. 각 메뉴의 내용(음식이름, 메뉴설명, 파일명)을 넣어 Document를 생성한다.

````python
# 문서 Load
import os
# 메뉴읽어 오기
menu_file_path = "data/restaurant_menu.txt"
with open(menu_file_path, "r", encoding="utf-8") as f:
    menu = f.read()
````

````python
# print(menu)
````

````python
# Split - 메뉴별로 분리하기 위해 직접 처리
from langchain_core.documents import Document
import re
from pprint import pprint

def create_documents(menu_all: str, file_name) -> list[Document]:
    """메뉴별로 분리 한뒤 Document로 변환. 변환된 Document들을 리스트로 반환
    metadata: 메뉴이름, 주요 재료, 가격, source(파일이름)
    """
    menu_list = menu_all.split("\n\n") # 개별 메뉴들로 나누기.
    menu_documents = [] # 각 메뉴 text -> Document 변환. Document들을 담을 List
    # page_content: 개별메뉴 text, metadata: 이름, 가격, 재료, 파일명
    for menu_item in menu_list:

        # 메뉴 이름 추출
        menu_name_pattern = r"(^\d+.\s)([가-힣a-zA-Z\d ]+)" # (번호. )(메뉴이름)
        menu_name_result = re.search(menu_name_pattern, menu_item)
        menu_name = menu_name_result.group(2).strip()  # 메뉴 이름 추출 # 1은 번호

        # 주요재료 추출 - text로 indexing.
        menu_ingredient_pattern = r"(?<=주요 재료: ).+"  # 긍정후방탐색. "주요 재료: " 는 찾을때만 사용하고 반환할때는 뺀다.
        ingredients = re.search(menu_ingredient_pattern, menu_item).group().strip()
    
        # 가격추출
        price_pattern = r"(?<=가격: )[\d,]+(?=원)" # (?<=가격: ) 후방탐색, (?=원) : 전방탐색
        price_result = re.search(price_pattern, menu_item)
        price = int(price_result.group().replace(",", ""))

        
        # 메뉴 이름을 제외한 나머지는 메뉴 설명으로 간주
        menu_doc = Document(
            page_content=menu_item,     # 메뉴 설명을 page_content에 넣는다. metadata는 나중에 따로 따로 payload filtering 할 수있기 때문에 
            metadata={
                "menu_name": menu_name, # 메뉴 이름
                "price": price,         # 가격
                "ingredients": ingredients, # 재료 리스트
                "source": file_name     # 메뉴가 저장된 파일 이름
            }
        )
        menu_documents.append(menu_doc)
    return menu_documents


menu_documents = create_documents(menu, menu_file_path)
print(len(menu_documents))
````

````python
# VectorDB 저장 및 VectorStore 생성
from langchain_qdrant import QdrantVectorStore
from langchain_openai import OpenAIEmbeddings
from qdrant_client import QdrantClient
from qdrant_client.http.models import Distance, VectorParams
from dotenv import load_dotenv

load_dotenv()

COLLECTION_NAME = "restaurant_menu"
VECTOR_SIZE = 1536  # OpenAIEmbeddings의 벡터 크기

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

client = QdrantClient(host="localhost", port=6333)

# collection이 있으면 삭제
if client.collection_exists(COLLECTION_NAME):
    client.delete_collection(COLLECTION_NAME)

# Collection 생성
client.create_collection(
    collection_name=COLLECTION_NAME,
    # 벡터 저장소에 대한 설정. - vector_config는 default이므로 이름을 안줘도 됨(default 로 잡힘) 이름을 줄 거면 {"이름":VectorParams()} 설정
    vectors_config=VectorParams(
        size=VECTOR_SIZE, 
        distance=Distance.COSINE
    )
)

vectorstore = QdrantVectorStore(
    client=client, # QdrantClient
    embedding=embeddings, # Embedding Model
    collection_name=COLLECTION_NAME # 연결할 Collection 지정.
)

ids = vectorstore.add_documents(menu_documents) # list[Document] -> upsert
print(f"저장된 문서 개수: {len(ids)}")
##################################################
# Retriever 생성
##################################################
menu_retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}  # 검색할 결과 개수
)

from langchain_core.runnables import ConfigurableField
retriever = menu_retriever.configurable_fields(
    search_type=ConfigurableField(
        id="search_type"
    ),
    search_kwargs=ConfigurableField(
        id="search_kwargs"
    ),
)
````

````python
#### @tool을 이용해서 정의
from pydantic import BaseModel, Field
class SearchMenuToolParameter(BaseModel):
    query:str = Field(..., description="의미기반 검색으로 메뉴를 찾을 query")
    top_k:int = Field(default=3, description="Vector DB에서 조회할 문서 개수. 기본값 3")

@tool(args_schema=SearchMenuToolParameter)
def search_menu_tool(query:str, top_k:int|None=3) -> str:
    """이 도구는 VectorStore에 저장된 restaurant menu를 검색하는 tool입니다. 음식 메뉴에 대한 정보가 필요할 때 사용합니다."""
    result_list = [] # 결과 문서들을 담을 list
    config = {
        "configurable": {
            "search_kwargs": {
                "k": top_k
            }
        }
    }
    docs = retriever.invoke(query, config=config)   
    for doc in docs:
        result_list.append(
            {
                "content":doc.page_content,
                "title":doc.metadata['menu_name'],
                "url": doc.metadata['source']
            }
        )
    if result_list:
        result = {"result": result_list}
    else:
        result = {"result": "검색된 메뉴가 없습니다."}
    
    return json.dumps(result, ensure_ascii=False) # JSON 문자열로 반환.
````

# create_agent() 를 이용한 Agent 구현

- create_agent()
  - 랭체인에서 제공하는 표준 에이전트 생성함수로 LLM이 도구를 사용해서 문제를 해결하는 기본구조를 기반으로 설계되었다.
    - Agent 생성 및 실행을 표준화, 단순화 하기 위헤 Langchain 1.0에서 도입 되었다.
    - 도구사용 기본 구조에 **middleware**를 이용해 다양한 시점에 다양한 동작이 실행될 수 있는 유연한 구조를 지원한다.

    ![기본구조](figures/langgraph_basic_flow.png)
  - 0.x 버전까지는 Agent 생성 함수는 
    1. model과 tool들을 넣어 Agent를 생성하고 
    2. 그것을 실행하는 Executor를 생성하는 방식으로
  - 여러 단계를 거쳐야 했다.
  - create_agent()는 이러한 단계를 1단계로 축약한 API이다.

- create_agent 설계 철학
    1. Agent를 구성하는 모든 실행단위는 Runnable 이다.
    2. Agent도 Runnable 이다.
    3. Agent는 하나의 표준 팩토리 함수(생성함수)로 통합한다.

## 주요 파라미터
- https://reference.langchain.com/python/langchain/agents/
- `model`
  - Agent에서 사용할 모델을 지정한다.
  - 문자열로 모델이름을 넣을 수도 있고 BaseChatModel(ex: ChatOpenAI) 타입의 객체를 전달 할 수도 있다.
- `tools`
  -  Agent가 사용할 외부 도구(tool) 목록(list)을 전달한다.
- `system_prompt`
  - Agent에 입력하는 시스템 프롬프트. 문자열 또는 `SystemMessage`객체로 입력한다.
  - 이 prompt는 모델 호출시 가장 먼저 입력되는 것으로 Agent의 행동 지침, 성격을 정의하는 데 사용된다.
- `middleware`
  - Agent 실행 흐름에서 특정 지점에서 동작을 가로채서 실행되는 middelware(중간처리기) 목록을 전달한다.
  - Agent실행 전/후, Model 호출 전/후 등 사전 정의된 hook에 맞춰 정의된 middleware들을 등록한다.
- `response_format`
  - Agent 최종 출력 결과 형식을 구조화하는 schema를 설정. 
- `checkpointer`
  - 에이전트 실행 중 상태를 저장하는 객체, 즉 단기 메모리 저장소을 설정한다.
- `store`
  - 지속성 저장소 즉 장기 메모리 저장소를 설정한다.

# LangChain에서 MCP 연동

## MCP란

- **MCP**(**Model Context Protocol**)는 Anthropic이 주도하여 2024년 말에 발표한 개방형 표준으로, LLM 애플리케이션이 외부 데이터 소스나 도구에 접근하는 방식을 표준화 하는 프로토콜이다.
  - Agent가 이용할 수 있는 tool을 구현하는 방법은 framework 마다 다르기 때문에 서로 호환 되지 않는 문제가 있다. 
  - MCP는 이런 문제를 해결하기 위해 만든 표준 프로토콜이다.

## MCP 아키텍처 구성 요소
- **MCP 호스트**
  -  AI 애플리케이션(예: LLM 채팅 애플리케이션)으로 MCP 클라이언트를 통해 여러 서버에 연결한다.

- **MCP 클라이언트**
  -  각 MCP 서버와의 연결을 유지하는 구성요소로, 서버로부터 도구/리소스/프롬프트를 가져와 LLM 모델에게 전달한다.

- **MCP 서버**
  -  외부 데이터 소스나 Tool들을 제공하는 프로그램이다. 서버는 도구, 리소스, 프롬프트를 노출하고 클라이언트와 JSON‑RPC 2.0 메시지로 통신한다.
  -  MCP는 서버 클라이언트 간의 **로컬 프로세스 간 통신을 위한 STDIO**와 **원격 서버 접속을 위한 Streamable HTTP** 두 가지 전송 방식을 지원한다
  -  **MCP 서버가 제공하는 것**
     -  **도구**(**Tools**): LLM이 호출할 수 있는 함수형 작업이다. 예를 들어 “queryDatabase”, “sendEmail” 등이 있으며 각 도구는 이름, 설명, 입력 스키마를 포함한다.
     -  **리소스**(**Resources**): 읽기 전용 데이터 소스로, 검색 색인·파일시스템·데이터베이스 등에서 컨텍스트를 제공한다.
     -  **프롬프트**(**Prompts**): LLM 프롬프트 템플릿이나 예시를 서버에서 제공해 AI 모델의 질의를 돕는다.

  - https://nodejs.org/ko/download

###  MultiServerMCPClient 
- **MCP 클라이언트 객체**로, LangChain에서 **하나 이상의 MCP 서버를 연결하고 Tool 목록을 가져오기 위해 사용**한다.
    - 여러 개의 MCP 서버를 **동시에 등록하고 연결**할 수 있다.
    - 서버 설정들을 **하나의 딕셔너리**(**Dictionary**)로 묶어 전달하며, **각 개별 서버 역시 Dictionary로 설정**한다.
    - MCP 클라이언트에 등록된 모든 서버는 이후 `get_tools()` 호출을 통해 서버가 제공하는 tool들을 **LangChain의 Tool 객체로 자동 변환해 Agent가 호출**(**사용**)할 수있게 한다.

#### 개별 서버 설정 Key 정의
- **`dict[str: 서버 별칭, dict: 서버 설정]`**

| 설정 Key        | 의미                                                                  | 사용 예                            |
| --------------- | --------------------------------------------------------------------- | ---------------------------------- |
| **`transport`** | 서버와 통신하는 방식                                                  | `"stdio"` 또는 `"streamable-http"` |
| **`command`**   | MCP 서버를 실행할 프로그램(프로세스)                                  | `"python"` 또는 `"npx"`            |
| **`args`**      | command에 전달할 세부 실행 인자를 **리스트(List)**에 문자열로 순서대로 배치 | `["-m", "mcp_server_time"]`  |
| **`env`**       | MCP 서버 실행시 필요한 환경변수 설정. (API Key)                       | `{"FIRECRAWL_API_KEY": os.getenv("FIRECRAWL_API_KEY")}`|

- **transport 값에 따른 동작 구분**
    - `"stdio"`: 로컬에서 실행되는 서버 연결
    - `"streamable-http"`: 원격 MCP 서버(서비스형) 연결
    
    ```json
        {
            "time": {
                "transport": "stdio",
                "command": "python",
                "args": ["-m", "mcp_server_time"]
            }
        
            "github": {
                "transport": "streamable-http",
                "url": "https://server.smithery.ai/github",
            }
        }
    ```

#### command + args 와 실행 명령어 관계

| 실제 실행 명령                    | 설정으로 표현되는 방식                                             |
| --------------------------- | -------------------------------------------------------- |
| `python -m mcp_server_time` | `"command": "python", "args": ["-m", "mcp_server_time"]` |
| `npx -y @org/server`        | `"command": "npx", "args": ["-y", "@org/server"]`        |

* **command는 실행 프로그램**, args는 **명령 뒤에 붙는 옵션을 순서 그대로 리스트에 나열**한다.

````python
%%writefile mcp_exam.py

from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent
from dotenv import load_dotenv
import os
import asyncio

load_dotenv()

async def main():
    # MultiServerMCPClient 객체 생성 -> MCP 서버와 연결 + mcp 툴들을 langchain tool로 변환.
    # 연결할 MCP 서버들 설정. dict[str:서버이름, dict]
    client = MultiServerMCPClient(
        {
            "time" : {
                "transport":"stdio",
                "command":"python",
                "args":["-m", "mcp_server_time"]
            },
            "filesystem": {
                "transport":"stdio",
                "command": "npx",
                "args": [
                    "-y",
                    "@modelcontextprotocol/server-filesystem",
                    os.getcwd()  # sever-filesystem이 사용할 수 있는 디렉토리.
                ]
            },
            "playwright": {
                "transport":"stdio",
                "command": "npx",
                "args": [
                    "@playwright/mcp@latest"
                ]
            }
        }
    )
    # Langchain Tool List로 변환
    tools = await client.get_tools()
    print("툴 개수:", len(tools))

    # Agent 생성
    agent = create_agent(
        model=ChatOpenAI(model="gpt-5.4-mini"),
        tools=tools,
        system_prompt="당신은 유능한 AI Assistant입니다. 필요한 경우 등록된 도구들을 이용해서 답변하세요"
    )
    print(">>>>>>>>> 종료하려면 !quit 를 입력하세요.<<<<<<<<")
    while True:
        query = input(">>>질문:")
        if query == '!quit':
            print(">>>종료<<<")
            break

        res = await agent.ainvoke({
            "messages":[
                ("human", query)
            ]
        })
        print("<<<답변:")
        print(res['messages'][-1].content)
        print("---------------------------------------")

if __name__ == "__main__":
    asyncio.run(main())
````
