# Model I/O

- **Model I/O는 언어 모델(LLM)에 입력을 전달하고, 생성된 출력을 처리하는 과정과 이를 구성하는 요소들을 의미한다.**
- LangChain에서는 주로 **프롬프트 템플릿**(**Prompt Template**), **언어 모델**(**Language Model**), **출력 파서**(**Output Parser**)를 사용하여 Model I/O를 구성한다.

## Model I/O 구성 요소

### 프롬프트 템플릿 (Prompt Template)

- LLM에 전달할 프롬프트를 생성하는 역할을 한다.
- 사용자 입력과 고정된 템플릿을 결합하여 일관된 형식의 프롬프트를 생성한다.
- 프롬프트를 구조화하여 모델이 원하는 방식으로 응답하도록 유도한다.

### 대형 언어 모델 (LLM - Large Language Model)

- 입력된 프롬프트를 기반으로 텍스트를 생성하거나 입력 내용을 분석하여 결과를 생성한다.
- OpenAI의 GPT, Anthropic의 Claude, Google의 Gemini 등이 대표적인 언어 모델이다.
- Langchain은 LLM 연동 API를 제공한다.

### 출력 파서 (Output Parser)

- 언어 모델의 출력을 원하는 형식으로 변환하는 역할을 한다.
- 텍스트 응답에서 필요한 정보를 추출하거나 JSON, Pydantic 객체 등 구조화된 데이터로 변환할 수 있다.
- 출력 데이터를 정제하고 후처리하여 다음 단계에서 활용하기 쉽도록 만든다.

# Chatbot 서비스 사용과 LLM API 사용의 차이

ChatGPT나 Claude 챗봇 서비스를 사용하는 것과 GPT 5.5 모델 또는 Opus 4.8 와 같은 LLM 모델을 API를 이용해 사용하는 것의 차이는 다음과 같이 정리할 수 있다.

| 구분        | Chatbot서비스 사용       | LLM API 사용 |
| --------- | ---------------- | ------------- |
| 대상        | 일반 사용자           | 개발자           |
| 목적        | AI를 바로 사용        | AI를 서비스에 내장   |
| 인터페이스     | 웹/앱              | 프로그래밍 코드      |
| 사용자 제어 수준     | 제한적       | 매우 높음         |
| 비용        | 구독료(Plus, Pro 등) | 사용량 기반 과금     |
| 개발 가능성    | 없음               | 가능            |
| 시스템 프롬프트  | 일부만 가능           | 완전 제어         |
| 외부 시스템 연동 | 제한적              | 자유롭게 가능       |

**ChatGPT와 같은 LLM 기반 Chatbot는 완성된 AI 서비스**이고, **OpenAI GPT 5.5 같은 API는 단순히 응답을 생성해 주는 LLM 모델을 이용해 AI 서비스를 만들기 위한 것**이다.

## Langchain을 이용한 OpenAI API 호출

- **ChatOpenAI**
    - chat (대화-채팅) 기반 모델 model.
    - Default 로 gpt-3.5-turbo 사용
    - llm 전달 입력과 llm 응답 출력 타입:  Message
- **Initializer 주요 파라미터**
    -  **model**
        -  사용할 openai 모델 지정
    -  **temperature**
        -  llm 모델의 출력 무작위성을 지정한다. 
        -  0 ~ 2 사이 실수를 설정하며 클 수록 무작위성이 커진다. 기본값: 0.7
        -  정확한 답변을 얻어야 하는 경우 작은 값을 창작을 해야 하는 경우 큰 값을 지정한다.
           -  정확한 답변: 0.0 ~ 0.3
           -  일반 챗봇: 0.5 ~ 0.8
           -  창작: 1.0 이상
    - **max_tokens**:
        - llm 모델이 응답할 최대 token 수.
        - gpt-5 모델은 추론과정을 거칠 수있고 이 경우 토큰을 소비한다. 그래서 max_tokens를 너무 작게 잡으면 응답이 안올 수있다.
    - **api_key**
        - OpenAI API key를 직접 입력해 생성시 사용.
        - API key가 환경변수에 설정 되있으면 생략한다. 
    - **OpenAI GPT-5 계열 모델 파라미터**
        - **reasoning_effort**
            - 모델의 추론 강도
            - None, "low", "medium", "high", "xlarge"
        - **verbosity**
            - 응답 길이 제어.
            - "low", "medium", "high"
-  메소드
    - **`invoke(message)`** : LLM에 프롬프트 메세지를 전달하며 LLM의 응답을 반환한다.

````python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model="gpt-5.4-nano",
    temperature=0.0,
)
response = model.invoke("OpenAI의 LLM 모델 종류는 뭐가 있나요?")
````

````python
response
````

````python
print(response.content)
````

````python
print(response.content)
````

````python
# streaming 방식으로 응답 받기.

response = model.stream("LangChain은 어떤 Framework인가요? 자세하게 설명해주세요.")
response
````

````python
for token in response:
    print(token.content, end="")
````

# Hugging Face 모델 사용

## Local 에 설치된 모델 사용
- HuggingFacePipeline 에 Model id를 전달해 Model객체를 생성한다.
- huggingface transformers 라이브러리를 이용해 model을 생성 한 뒤 HuggingFacePipeline 에 넣어 생성한다.
- 모델이 local에 없는 경우 다운로드 받는다.

### HuggingFace 모델을 사용하기 위한 package 설치
```bash
pip install langchain-huggingface transformers torch -qU
```
- nvidia GPU가 있는 경우 `torch cuda`  버전을 먼저 설치하고 `langchain-huggingface`를 설치 해야 한다. 아니면 `torch cpu` 버전이 설치된다.

````python
from langchain_huggingface.llms import HuggingFacePipeline

model_id = "Qwen/Qwen2.5-0.5B-Instruct"
model = HuggingFacePipeline.from_model_id(
    task="text-generation", model_id=model_id
)
````

````python
res = model.invoke("HuggingFace를 소개해줘.")
````

````python
print(res)
````

````python
message = [
    {"role":"user", "content":"서울에서 추천하는 여행지 3곳을 알려줘."}
]
res = model.invoke(message)
````

````python
print(res)
````

````python
gpt = ChatOpenAI(model="gpt-5.4-mini")
res_gpt = gpt.invoke(message)
````

````python
print(res_gpt.content)
````
