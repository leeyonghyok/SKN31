# 프롬프트 엔지니어링(Prompt Engineering)

## 프롬프트(Prompt)

- 프롬프트(Prompt)는 생성형 인공지능 모델에게 작업을 요청하거나 지시하기 위해 입력하는 정보이다.
- 일반적으로 자연어로 작성되지만, 이미지, 음성, 동영상 등의 다양한 형태로 입력될 수도 있다.
- 생성형 인공지능 모델은 프롬프트를 바탕으로 텍스트, 이미지, 코드 등 다양한 결과를 생성한다.

## 프롬프트 엔지니어링이란

- **프롬프트 엔지니어링**은 LLM이 원하는 결과를 안정적으로 생성하도록 **지시문, 입력 정보, 출력 형식, 예시, 제약 조건을 설계하는 작업**이다.
    - OpenAI는 프롬프트 엔지니어링 정의:  "모델이 요구사항을 만족하는 콘텐츠를 **일관되게 생성하도록 효과적인 지시를 작성**하는 과정"
    - LLM 출력은 비결정적이지만, 프롬프트에 역할, 예시, 평가 등을 사용하여 구체화 하면 더 일관된 결과를 얻을 수 있다.
    - 예를 들어 단순히 다음처럼 요청하면 결과가 들쭉날쭉하고 장황해 질 수 있다.

        ```text
        프롬프트 엔지니어링 설명해줘.
        ```

        반면 다음처럼 요청하면 더 안정적인 결과를 얻을 수 있다.

        ```text
        너는 AI를 초보자에게 가르치는 강사이다.
        프롬프트 엔지니어링을 초보자에게 설명하라.

        조건:
        - 먼저 한 문장으로 정의한다.
        - 그다음 원칙 5개를 표로 정리한다.
        - 마지막에 좋은 프롬프트와 나쁜 프롬프트 예시를 비교한다.
        - 한국어로 작성한다.
        ```

    -  프롬프트 엔지니어링은 **AI에게 막연히 질문하는 기술이 아니라, 원하는 작업을 수행하도록 명세서를 작성하는 기술** 이다.


## 프롬프트 엔지니어링의 발전
  - 프롬프트 엔지니어링 -> 컨텍스트 엔지니어링 -> 하네스 엔지니어링
  - **프롬프트 엔지니어링은 모델에게 '무엇을 하라고 지시할 것인가'**, **컨텍스트 엔지니어링은 '모델에게 어떤 정보를 제공할 것인가'**, **하네스 엔지니어링은 '모델이 실제로 일하게 만드는 실행 환경을 어떻게 설계할 것인가'** 를 말한다.


| 구분         | 핵심 질문                         | 주요 대상                        |
| ---------- | ----------------------------- | ---------------------------- |
| 프롬프트 엔지니어링 | 모델에게 어떻게 지시할 것인가?             | 지시문, 역할, 형식, 예시              |
| 컨텍스트 엔지니어링 | 모델에게 어떤 정보를 제공할 것인가?        | RAG, 검색결과, 도구 결과, 메모리, 대화 이력       |
| 하네스 엔지니어링  | 모델을 어떤 시스템 안에서 어떻게 실행·검증·통제할 것인가? | 가드레일(Guardrails), 에이전트 피드백 루프, 코딩표준, 아키텍처 규칙, 도구, 권한, 테스트, 로깅, 검증 |


## 프롬프트 엔지니어링의 원칙

### 1. 목표를 명확히 하고 내용을 구체적으로 작성한다. (명확성, 구체성)

좋은 프롬프트는 "무엇을 해줘"가 아니라 **무엇을, 누구를 위해, 어떤 수준으로, 어떤 형식으로** 해달라는 요청을 포함한다.
- 프롬프트는 생성형 인공지능 모델이 수행해야 할 작업을 명확하고 구체적으로 설명해야 한다.
- 모호하거나 불완전한 지시를 제공하면 모델은 사용자의 의도를 정확하게 파악하기 어렵고, 기대와 다른 결과를 생성할 수 있다.
- 따라서 수행할 작업, 원하는 결과의 형식, 대상 독자, 분량, 제약 조건 등을 가능한 한 구체적으로 제시하는 것이 좋다.

나쁜 예:

```text
머신러닝 설명해줘.
```

좋은 예:

```text
머신러닝을 처음 배우는 비전공자를 대상으로 설명해줘.
머신러닝의 정의, 지도학습/비지도학습 차이, 실제 활용 예시를 포함하고, 마지막에 5문장으로 요약해줘.
```

Anthropic도 프롬프트를 개선하기 전에 먼저 **성공 기준을 명확히 정의하고, 그 기준을 경험적으로 테스트할 방법을 갖추는 것**을 권장한다.


### 2. 역할(Persona)을 지정한다

모델에게 역할(Persona)을 주면 답변의 관점과 깊이가 달라진다.
역할 지정은 단순한 "연기"가 아니라, **답변 기준, 용어 선택, 예시의 수준, 판단 기준을 좁히는 장치** 이다.

예1:

```text
당신은 10년 차 경력의 숙련된 UX 디자이너입니다. 
사용자 중심 디자인과 데이터 기반 의사결정에 강점이 있습니다.
신규 모바일 쇼핑 앱의 첫 화면 구성안을 제안해야 합니다.

...

위 역할을 바탕으로 화면 구성을 제안해주세요.
```

예2:
```text
당신은 10년 차 전문 헬스 트레이너이다.
체지방을 줄이고 근육량을 늘리고 싶은 30대 직장인 남성을 위한 1주일 운동 스케쥴을 작성해줘.
...
```

### 3. 출력 형식을 명확히 지정한다

LLM은 기본적으로 자유형 텍스트를 생성하므로, 원하는 결과 형식이 있으면 명시해야 한다. API나 자동화 시스템에서는 특히 중요하다.

예:

```text
...

결과를 JSON  형식으로 작성해줘.

{
  "person": "김철수",
  "task": "보고서 제출",
  "due_date": "2025-12-31"
}

```


### 4. 필요한 배경 정보(Context-맥락 정보)를 제공한다

- 모델은 사용자의 상황을 자동으로 알지 못한다. 따라서 답변에 필요한 배경 정보가 있다면 넣어야 한다.
- 모델 답변의 할루시네이션을 방지하기 위해 다양한 형태의 정보를 Context에 포함하여 제공할 수 있다. 이런 정보에는 사내 업무 문서, 웹 검색정보, 데이터베이스 등에서 조회한 관련 정보들이 있을 수 있다.


```text
우리는 월 20%의 고객 이탈률을 겪고 있는 구독 기반 온라인 교육 서비스이다.   
주요 고객층은 30대 직장인이고, 이탈 주요 원인은 콘텐츠 부족과 동기 저하이다.   
이탈률을 줄이기 위한 실질적인 전략을 제안해줘.
```

### 5. 예시를 제공한다 

모델에게 원하는 입출력 예시를 주면 결과가 안정된다.
예시의 개수에 따라 **Zero shot Learning**, **One shot Learning**, **Few shot Learning** 이 있다.

```text

예시:
질문: "과적합이란 무엇인가요?"
답변:
    - 정의: 학습 데이터에는 잘 맞지만 새로운 데이터에는 성능이 낮아지는 현상
    - 비유: 문제집 답을 외웠지만 새로운 문제를 못 푸는 상황
    - 예시: 훈련 정확도 99%, 테스트 정확도 70%
```

### 6. 프롬프트 구성을 구조화한다.
긴 프롬프트는 역할(Role), 지시(Instruction), 맥락(Context), 출력 형식(Output Format) 등을 구획으로 나눈다. 나눌 때는 **태그**(예: `<context>, <instructions>`)나 **Markdown**(`#`, `##`)를 이용한다.

```xml
<instructions>
다음 뉴스 기사를 읽고 핵심 내용을 요약하라.
</instructions>

<context>
사용자는 바쁜 직장인으로, 전체 기사를 읽을 시간이 없다. 핵심 정보만 빠르게 파악하고 싶어한다.
</context>

<input_data>
[기사 원문]
오늘 오전, 서울 강남구 일대에 시간당 80mm가 넘는 집중호우가 내려 도로 일부가 침수되고, 차량 통제가 이루어졌다. 기상청은 이 같은 집중호우가 오후까지 계속될 것으로 전망하면서 시민들의 외출 자제를 당부했다. 서울시 역시 재난 문자를 통해 대중교통 이용을 권고하고 있다. ...
</input_data>

<output_format>
한 문단(3줄 이내)으로 요약하라. 일반인이 쉽게 이해할 수 있는 문장을 사용하라.
</output_format>

<constraints>
- 전문 용어에는 반드시 쉬운 설명을 덧붙일 것
- 분량은 500자 이내 한다.
- 한국어로 작성한다.
</constraints>
```


### 7. 제약 조건을 명시한다

- 모델이 하지 말아야 할 것도 명확히 써야 한다.
- 특히 교육, 법률, 의료, 투자, 기업 보고서처럼 정확성이 중요한 영역에서는 **출처, 기준 시점, 불확실성, 검증 절차**를 프롬프트에 포함하는 것이 좋다.

예:

```text
주의:
- 모르는 내용은 추측하지 말고 모른다고 말한다.
- 현재 날짜 2XXX 년 10월 20일을 기준으로 최신정보를 제공한다.
- 코드는 Python 3.12 기준으로 작성한다.
```

### 8. 프롬프트를 점진적으로 개선한다.

실무에서는 프롬프트를 한 번 작성하고 끝내지 않는다. 원하는 결과가 안정적으로 나오는지 반복적으로 평가하고 튜닝해야 한다.
프롬프트를 평가할 때는 평가에 사용할 LLM 모델의 버전 고정하여 프롬프트 수정 후 결과가 달라졌을 때, 그 변화가 모델 변경 때문인지 프롬프트 변경 때문인지 구분할 수 있도록 한다.
테스트 질문, 기대 답변, 평가 기준을 미리 정해두고 프롬프트 성능을 측정하는 평가 방식을 구축한다. 이를 통해 프롬프트를 수정할 때마다 정확도, 형식 준수 여부, 누락 여부, 환각 여부 등의 변화를 모니터링할 수 있다.


# 프롬프트 템플릿 (Prompt Template)

프롬프트 템플릿은 언어 모델(Large Language Model, LLM)에 입력할 **프롬프트를 생성하는 재사용 가능한 템플릿**이다. 프롬프트 작성 시 변수를 포함하여 템플릿화함으로써, 다양한 입력값에 유연하게 대응할 수 있으며, 일관성 있는 결과를 얻는 데 유리하다.
프롬프트는 문자열(string)로 작성되므로 Python의 f-string을 사용해 동적으로 생성할 수 있다. 또한 LangChain은 다양한 PromptTemplate 클래스를 제공하여 변수 치환, 프롬프트 재사용, 체계적인 프롬프트 관리를 지원한다.

## 프롬프트 템플릿의 목적

* 재사용성: 입력값이 달라져도 동일한 프롬프트 구조를 재사용할 수 있다.
* 유지보수성 향상: 프롬프트 구조와 변수 값을 분리하여 관리할 수 있으므로 수정과 관리가 용이하다.
* 자동화 지원: 사용자 입력, 데이터베이스 조회 결과, 웹 검색 결과 등 다양한 데이터를 템플릿에 주입하여 프롬프트를 자동으로 생성할 수 있다.
* 일관성 확보: 동일한 형식의 프롬프트를 지속적으로 생성하여 모델의 응답 품질을 보다 안정적으로 유지할 수 있다.

## Langchain의 주요 Prompt Template

1. PromptTemplate

* **단일 텍스트(string) 형태의 프롬프트를 생성**하기 위한 가장 기본적인 템플릿 클래스이다.
* 특징
    * 문자열 기반 프롬프트 생성에 사용된다.
    * 변수(Placeholder)는 중괄호 {}로 표시한다.
      * `{}`는 변수를 표시하는 구분자로 사용된다. 그래서 프롬프트에 중괄호 자체를 문자열로 포함하려면 `{{`와 `}}`를 사용하여 이스케이프(escape)해야 한다.
    * 변수(placeholder)에 전달되는 입력값에 따라 프롬프트를 동적으로 생성할 수 있다.
    * 주로 텍스트 완성(Text Completion) 방식이나 단순 지시형 프롬프트 작성에 적합하다.

1. ChatPromptTemplate

* 대화형(Chat) 언어 모델을 위한 프롬프트 템플릿 클래스이다.
* 시스템(System), 사용자(Human), AI(AI) 등의 역할(Role)을 가진 메시지들을 조합하여 프롬프트를 구성할 수 있다.
* 특징
    * 역할(Role) 기반 메시지를 정의할 수 있다.
    * 시스템 메시지(SystemMessage), 사용자 메시지(HumanMessage), AI 메시지(AIMessage) 등을 하나의 프롬프트로 구성할 수 있다.
    * 대화 이력이나 컨텍스트를 구조적으로 관리하기 쉽다.
    * GPT 계열과 같은 채팅 모델에서 가장 많이 사용되는 프롬프트 템플릿이다.

- 최근의 GPT, Claude, Gemini, Qwen과 같은 대부분의 LLM은 채팅(Chat) 인터페이스를 기반으로 동작하므로, PromptTemplate보다 ChatPromptTemplate이 더 널리 사용된다.

## 프롬프트 템플릿 사용법

- 전체 템플릿은 문자열(string)로 정의하며, 변수(placeholder)는 `{변수명}` 형식으로 지정한다. 템플릿이 실행될 때 변수는 전달된 실제 값으로 대체된다.
  - 예시: `"{country}의 수도는 어디인가요?"`
- 프롬프트에 중괄호(`{, }`) 자체를 포함하려면 `{{` 와 `}}`를 사용하여 이스케이프(escape)해야 한다.
  - 예시: `{{"이름": "{name}"}}`
- 생성한 문자열 템플릿을 이용하여 Prompt Template 객체를 생성한다.

### 공통 메소드

프롬프트 템플릿 클래스에서 공통적으로 사용하는 주요 메소드는 다음과 같다.

#### 템플릿 생성 메소드

- **from_template()**
  - 문자열 기반 템플릿을 생성한다.
  - `PromptTemplate`과 `ChatPromptTemplate` 모두에서 사용할 수 있다.

- **from_messages()**
  - 메시지 객체를 이용하여 대화형 템플릿을 생성한다.
  - `ChatPromptTemplate`에서 사용한다.

#### 프롬프트 생성 메소드

- **format(변수=값, ...)**
  - 변수에 값을 대입하여 최종 프롬프트 문자열을 생성한다.

- **format_messages(변수=값, ...)**
  - 메시지 리스트 형식의 프롬프트를 생성한다.
  - `ChatPromptTemplate`에서 사용한다.

- **invoke(dict)**
  - 변수와 값을 딕셔너리 형태로 전달하여 프롬프트를 생성한다. 생성된 프롬프트는 `PromptValue` 객체로 반환된다.
  - 생성된 결과는 Runnable 체인의 다음 단계 입력으로 사용할 수 있다.

> ##### `invoke()` 메소드
>
> - invoke()는 LangChain의 핵심 인터페이스인 Runnable에서 제공하는 공통 실행 메소드이다.
> - Runnable은 LangChain의 실행 가능한 객체들이 따르는 공통 인터페이스이다.
> - PromptTemplate, ChatModel, OutputParser, Chain 등은 모두 Runnable 인터페이스를 구현한다.
> - invoke()는 입력을 받아 처리한 뒤 결과를 반환한다.
> - 여러 Runnable 객체를 연결하여 체인(Chain)을 구성할 수 있으며, 각 단계의 출력은 다음 단계의 입력으로 전달된다.

## MessagesPlaceholder

- `MessagesPlaceholder`는 `ChatPromptTemplate` 내에서 여러 개의 메시지가 삽입될 위치를 지정하는 프롬프트 템플릿 요소이다.
- 주로 대화 이력(Chat History), 퓨샷 예제(Few-shot Examples), 에이전트 중간 처리 결과 등을 프롬프트에 동적으로 추가할 때 사용한다.
- 일반 변수({variable})가 문자열이나 단일 값을 삽입하는 데 사용된다면, `MessagesPlaceholder`는 여러 개의 메시지(Message 객체 목록)를 삽입하는 데 사용된다.
- `MessagesPlaceholder`는 `ChatPromptTemplate`에서만 사용할 수 있으며, 일반 `PromptTemplate`에서는 사용할 수 없다.

### Initializer의 주요 파라미터

1. **variable_name** (str)
   - 프롬프트에서 참조할 변수명을 지정한다.
   - invoke() 실행 시 동일한 이름의 메시지 리스트를 전달해야 한다.
1. **optional** (bool, 기본값: False)
   - True로 설정하면 해당 변수에 메시지가 전달되지 않아도 오류를 발생시키지 않는다.
   - 주로 대화 이력이 없을 수 있는 경우에 사용한다.
1. **n_messages** (int)
   - 지정된 경우 최근 N개의 메시지만 프롬프트에 포함한다.
   - 긴 대화 이력을 제한하여 토큰 사용량을 줄이는 데 활용할 수 있다.



````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

prompt = ChatPromptTemplate(
    messages=[
        ("system", "당신은 수학 전문 Assistant입니다."),
        MessagesPlaceholder(variable_name="history", optional=True),
        ("human", "{query}")
    ]
)

# 지금까지의 대화 내역
chat_history = [ 
    ("human", "5 + 2의 결과는?"),
    ("ai", "5 + 2 = 7"),
    ("human", "10 + 22의 결과는?"),
    ("ai", "10 + 22 = 32")
]

query = prompt.invoke({
    'history': chat_history,
    "query":"위 결과에 4제곱을 하면 얼마인가요?"
})
query.messages
````

````python
res = model.invoke(query)
res.content
````

````python
system_msg = ("system", "당신은 유능한 assistant입니다. 답변은 10단어 이내로 해주세요.")
prompt = ChatPromptTemplate(
    messages=[
        system_msg, 
        MessagesPlaceholder(variable_name="history", optional=True),
        ("user", "{query}")
    ]
)
user_input = input("질문:")

chat_history = [system_msg]

while True:
    if user_input.strip() == "quit":
        break
    query = prompt.invoke({"query":user_input, "history":chat_history})
    res = model.invoke(query)

    print("질문:", user_input)
    print("답변:", res.content)

    chat_history.append(("human", user_input))
    chat_history.append(("ai", res.content))

    user_input = input("질문:")
````

# 멀티모달(Multimodal) 프롬프트

## 모달리티(Modality)
- 모달리티(Modality)란 정보를 표현하거나 전달하는 방식 또는 형태를 의미한다. 대표적인 모달리티로는 텍스트(Text), 이미지(Image), 오디오(Audio), 비디오(Video) 등이 있다.
  - "오늘 비가 온다"라는 정보를 표현하는 방법으로
    * **텍스트**: "오늘 비가 옵니다."
    * **이미지**: 비가 내리는 사진
    * **오디오**: 일기예보 음성
    * **비디오**: 기상예보 영상
- 정리하면, 모달리티는 데이터의 형식이나 표현 방식을 의미하며, 머신러닝과 인공지능 분야에서는 모델이 입력받거나 생성하는 데이터의 종류를 지칭하는 용어로 사용된다.

## 멀티모달 AI

- **멀티모달(Multimodal) AI**란 여러 형식의 입력 데이터(모달리티)를 동시에 처리하고 이해할 수 있는 인공지능 시스템을 의미한다.  
  - 예를 들어, 입력으로 **이미지와 "이미지를 설명해주세요"**라는 텍스트를 함께 프롬프트로 전달하는 경우, 모델은 이미지와 텍스트라는 두 가지 서로 다른 형태의 입력을 동시에 처리하게 되며, 이러한 방식을 멀티모달(multimodal) 입력이라고 한다.
  - 단일 모달리티만 처리하는 것은 유니모달(Unimodal) AI라고 한다.  
- 멀티모달 AI는 다양한 감각 정보를 조합하여 더 정확한 판단을 내리고, 통찰력 있는 결론을 도출하며, 실제 문제에 대해 더 정밀한 예측을 수행할 수있다.

### 다양한 입력 모달리티의 예

- 텍스트: 자연어, 문서, 코드
- 이미지: 사진, 차트
- 오디오: 음성, 음악, 소리
- 비디오: 동영상, 애니메이션
- 기타: 수학 방정식, 센서 데이터 등

## 멀티모달 프롬프트

- **멀티모달 프롬프트**란 인공지능 언어 모델(LLM, Large Language Model)에 입력할 때 텍스트뿐 아니라 이미지나 오디오와 같은 다양한 유형의 데이터를 함께 제공하는 방식을 말한다.
- 이를 통해 LLM은 다양한 감각 정보를 조합하여 더 깊이 있고 풍부한 응답을 생성할 수 있다. 예를 들어, 사용자가 사진과 함께 질문을 하면 모델은 사진의 내용을 이해하고 텍스트로 된 설명을 결합하여 답변할 수 있다.
  
### 멀티모달 프롬프트 예
- **이미지 + 텍스트**: 사진을 업로드하고 "이 이미지에서 무엇을 개선할 수 있나요?" 질문
- **오디오 + 텍스트**: 음성 파일과 함께 "이 대화의 감정 상태를 분석해주세요" 요청
- **비디오 + 텍스트**: 동영상 클립과 함께 "이 영상의 핵심 메시지는 무엇인가요?" 문의

# Langchain 멀티모달 프롬프트 지원
- LangChain은 다양한 인공지능 모델과 연동하여 텍스트뿐만 아니라 이미지, 오디오 등 여러 종류의 입력 데이터를 처리할 수 있도록 멀티모달 프롬프트 기능을 지원한다.
  
## 지원 가능 모델 확인
- Langchain으로 멀티모달 입력 기능을 사용하려면, 연동하려는 LLM 모델이 모달리티 입력을 지원해야 한다.
- [Langchain 지원 LLM 모델](https://python.langchain.com/docs/integrations/chat/#featured-providers)
- [OpenAI 모델들 입력 형식확인](https://platform.openai.com/docs/models)

## LangChain에서 멀티모달 입력 구현 방법
- HumanMessage 객체의 content에 텍스트 외에도 이미지, 오디오등 다양한 데이터를 함께 포함시킬 수 있다.
- **텍스트**: 일반 문자열 형태로 입력
- **이미지/오디오**: 다음 두 가지 방식으로 입력할 수 있다.
    - **URL 지정**: 인터넷에 업로드된 이미지나 오디오의 URL을 제공
    - **Base64 인코딩**: 파일 데이터를 base64 형식으로 인코딩하여 직접 전달

> **Base64 인코딩**은 이진 데이터를 텍스트로 안전하게 변환하기 위한 인코딩 방식이다.   
> 주로 이메일, URL, JSON, HTML 등 텍스트 기반 시스템에서 이진 데이터를 전송하거나 저장할 때 사용한다.
> Base64는 이진 파일을 64개의 문자(`영문 대소문자, 숫자, +, /`)로 인코딩한다. ASCII 문자들만 사용하기 때문에 대부분의 시템에서 안전하게 처리할 수있다.

````python
import base64

def encode_to_base64(src_path:str)->str:
    """src_path 경로의 binary 파일을 읽어서 base64 로 인코딩 한 뒤 반환. """
    with open(src_path, 'rb') as fi:
        encoded_bytes = base64.b64encode(fi.read())
        encoded_str = encoded_bytes.decode('utf-8')
        return encoded_str
    
````
