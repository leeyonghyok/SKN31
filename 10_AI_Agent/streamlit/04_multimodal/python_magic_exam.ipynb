# python-magic

- 파일 타입 식별 라이브러리로 파일의 확장자가 아니라 실제 바이너리 시그니처(표시)를 읽어서 "이 파일이 PDF인지, PNG인지, ZIP인지" 등을 판별한다.
- 사용법
  - **파일경로로 부터 확인**
    - `file_type = magic.from_file(str: 확인할 파일 경로, mime:bool)`
      - `mime=False`(기본): 사람이 읽을 수 있는 상세 설명 반환
      - `mime=True`: 마임타입 형식 문자열 반환
  - **bytes로 부터 확인**
    - `file_type = magic.from_buffer(data:bytes, mime: bool)`

## 설치
- **WINDOWNS**
    - `pip install python-magic-bin`

- **MacOS**
  - libmagic 설치 필요
    ```bash
        brew install libmagic
        pip install python-magic
    ```

````python
import magic
#######################################################################
# 파일 경로로 부터 조회
#######################################################################
file_path = "data/README.md"
file_path = "data/news.png"
file_path = "data/config.py"
file_path = "data/news.dat" # news.png의 확장자 변경한 것
file_path = "data/dogs.jpg"
file_path = "data/doc.pdf"

file_type = magic.from_file(file_path, mime=True)
print(file_type)
````

````python
#######################################################################
# bytes 부터 조회
#######################################################################
file_path = "data/README.md"
file_path = "data/news.png"
file_path = "data/config.py"
file_path = "data/news.dat" # news.png의 확장자 변경한 것
file_path = "data/dogs.jpg"
file_path = "data/doc.pdf"

with open(file_path, "rb") as fi:
    bytes_data = fi.read()

file_type = magic.from_buffer(bytes_data, mime=True)
file_type
````
