```python
import random

random.seed(13472)

stu_list = [
    "고현아",
    "김가율",
    "김동민",
    "김봉남",
    "김세희",
    "김재원",
    "김효민",
    "박동관",
    "박연아",
    "박하린",
    "안영선",
    "안혁진",
    "오형호",
    "유진영",
    "이영창",
    "이용혁",
    "이재일",
    "이한수",
    "전서연",
    "정형섭"
]
print(len(stu_list))
```

```python
num = 4
random.shuffle(stu_list)
result = [stu_list[i * num : (i + 1) * num] for i in range((len(stu_list) + num - 1) // num )] 

for idx, r in enumerate(result, start=1):
    print(f"{idx}조: ", end="")
    for name in r:
        print(name, end=',  ')
    print()
```
