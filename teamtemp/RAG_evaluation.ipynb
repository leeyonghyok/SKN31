## 0. Qdrant / Neo4j 테스트셋 로드 및 병합
- 각 노트북(`RAG_evaluation_Qdrant_1.ipynb`, `RAG_evaluation_Neo4j.ipynb`)에서 `eval_df.to_pickle(...)`로 저장해둔 테스트셋을 불러와 병합합니다.
- **이 셀을 가장 먼저 실행**해야 아래 셀들에서 `eval_df`를 참조할 수 있습니다 (지금 노트북은 이 셀이 없어서 `NameError: name 'eval_df' is not defined`가 발생했습니다).

```python
import pandas as pd
import os

# 노트북이 프로젝트 루트(SKN31-3rd-2Team)에 있으므로 상대경로로 충분합니다.
SAVE_DIR = "eval_results"

qdrant_path = os.path.join(SAVE_DIR, "eval_testset_qdrant.pkl")
neo4j_path = os.path.join(SAVE_DIR, "eval_testset_neo4j.pkl")

assert os.path.exists(qdrant_path), f"파일이 없습니다: {os.path.abspath(qdrant_path)}"
assert os.path.exists(neo4j_path), f"파일이 없습니다: {os.path.abspath(neo4j_path)}"

df_qdrant = pd.read_pickle(qdrant_path)
df_neo4j = pd.read_pickle(neo4j_path)

# 컬럼 구성이 동일한지 확인 (user_input, reference_contexts, reference, synthesizer_name, source 등)
assert set(df_qdrant.columns) == set(df_neo4j.columns), (
    f"두 테스트셋의 컬럼이 다릅니다.\nqdrant: {list(df_qdrant.columns)}\nneo4j: {list(df_neo4j.columns)}"
)

eval_df = pd.concat([df_qdrant, df_neo4j], ignore_index=True)
print(eval_df.shape)
print(eval_df["source"].value_counts())
eval_df
```

## 1. 병합된 user_input을 에이전트(`LangGraphChatbot`)에 넣어 response / retrieved_contexts 생성

```python
import sys, os, re, uuid

# 노트북 실행 위치(cwd) = 프로젝트 루트이므로, run_chatbot.py/tools.py를 바로 import 가능
# (혹시 다른 위치에서 커널을 띄웠을 경우를 대비해 루트를 sys.path에 명시적으로 추가)
PROJECT_ROOT = os.getcwd()
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

from dotenv import load_dotenv
load_dotenv(os.path.join(PROJECT_ROOT, ".env"))

from run_chatbot import LangGraphChatbot

bot = LangGraphChatbot(verbose=False)

def split_references(references: list[str]) -> list[str]:
    contexts = []
    for ref in references:
        chunks = re.split(r"--- \[참조 (?:문서|조문) \d+\] ---", ref)
        for c in chunks:
            c = c.strip()
            if c:
                contexts.append(c)
    return contexts

response_list = []
retrieved_context_list = []

for i, user_input in enumerate(eval_df["user_input"], start=1):
    thread_id = f"eval-{uuid.uuid4()}"
    references, answer = bot.ask(user_query=user_input, thread_id=thread_id)
    contexts = split_references(references)

    if not contexts:
        print(f"[경고] {i}번째 질문에서 검색된 context가 없습니다: {user_input}")

    response_list.append(answer)
    retrieved_context_list.append(contexts)

    if i % 5 == 0:
        print(f"진행: {i}/{len(eval_df)}")

eval_df["response"] = response_list
eval_df["retrieved_contexts"] = retrieved_context_list
eval_df
```

## 2. RAGAS 통합 평가

```python
from ragas import EvaluationDataset, evaluate
from ragas.metrics import LLMContextRecall, LLMContextPrecisionWithReference, Faithfulness, AnswerRelevancy
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper
from langchain_openai import ChatOpenAI, OpenAIEmbeddings

eval_dataset = EvaluationDataset.from_pandas(
    eval_df[["user_input", "retrieved_contexts", "response", "reference"]]
)

eval_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4.1"))
eval_embeddings = LangchainEmbeddingsWrapper(OpenAIEmbeddings(model="text-embedding-3-large"))

metrics = [
    LLMContextRecall(llm=eval_llm),
    LLMContextPrecisionWithReference(llm=eval_llm),
    Faithfulness(llm=eval_llm),
    AnswerRelevancy(llm=eval_llm, embeddings=eval_embeddings),
]

eval_result = evaluate(dataset=eval_dataset, metrics=metrics)
result_df = eval_result.to_pandas()
result_df["source"] = eval_df["source"].values  # 출처 컬럼 다시 붙이기
result_df
```

## 3. 출처(Qdrant / Neo4j)별 평균 점수 비교

```python
metric_cols = ["context_recall", "llm_context_precision_with_reference", "faithfulness", "answer_relevancy"]

summary = result_df.groupby("source")[metric_cols].mean()
summary.loc["total"] = result_df[metric_cols].mean()
summary
```

## 4. (선택) 결과 저장

```python
result_path = os.path.join(SAVE_DIR, "ragas_eval_result_merged.csv")
result_df.to_csv(result_path, index=False, encoding="utf-8-sig")
print(f"저장 완료: {os.path.abspath(result_path)}")
```
